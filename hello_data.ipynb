```python
# === Test de la stack data ===
import pandas as pd
import matplotlib.pyplot as plt

# 1. Création d'un petit jeu de données
data = {
    'Mois': ['Jan', 'Fév', 'Mar', 'Avr', 'Mai', 'Juin'],
    'Ventes': [120, 145, 98, 167, 189, 210],
    'Retours': [12, 18, 8, 22, 15, 19]
}
df = pd.DataFrame(data)

# 2. Affichage du DataFrame
print("Aperçu du DataFrame :")
df
```

```python
# === Visualisation des ventes et retours ===
import matplotlib.pyplot as plt

# Création de la figure et des axes
fig, ax = plt.subplots(figsize=(10, 5))

# Tracé des deux courbes
ax.plot(df['Mois'], df['Ventes'], marker='o', linewidth=2, label='Ventes', color='#1F4E79')
ax.plot(df['Mois'], df['Retours'], marker='s', linewidth=2, label='Retours', color='#C9A227')

# Habillage du graphique
ax.set_title('Évolution mensuelle — Ventes vs Retours', fontsize=14, fontweight='bold')
ax.set_xlabel('Mois')
ax.set_ylabel('Nombre')
ax.legend()
ax.grid(True, alpha=0.3)

# Affichage
plt.tight_layout()
plt.show()
```
