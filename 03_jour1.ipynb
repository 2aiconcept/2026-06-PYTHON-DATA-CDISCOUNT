### 🛠️ Mini-TP — Calcul d'un prix promo TTC

Un produit est vendu **39,90 € HT**.  
Pour la promo de fin d'année, on applique :
1. Une **remise de 15 %** sur le prix HT
2. Puis la **TVA de 20 %** sur le prix remisé

**Objectif** : afficher exactement cette ligne (avec 2 décimales) :

Prix initial : 39.90 € HT — Après remise de 15 % : XX.XX € HT — Prix final TTC : XX.XX €



**Contraintes** :
- Utilise des variables (`prix_ht`, `taux_remise`, `taux_tva`)
- Utilise les opérateurs arithmétiques
- Affiche avec une **f-string** et le format `.2f` pour les décimales


````python
prix_ht = 39.90
taux_remise = 0.15
taux_tva = 0.20

print(prix_ht)
print(taux_remise)
print(taux_tva)

prix_remise = prix_ht * (1 - taux_remise)
prix_ttc = prix_remise * (1 + taux_tva)

print(prix_remise)
print(prix_ttc)

print(f"Prix initial : {prix_ht:.2f} € HT -- Après remise de {taux_remise:.0%} : {prix_remise:.2f} € -- Prix final TTC : {prix_ttc:.2f} €")

nom_produit = "Smartphone X12"
print(f"""
Bonus f-string multiligne
=== Rapport produit ===
Nom : {nom_produit}
Prix HT : {prix_ht:.2f} €
Prix après remise de {taux_remise:.0%} : {prix_remise:.2f} €
TVA {taux_tva:.0%} : {prix_ttc - prix_remise:.2f} €
Prix TTC : {prix_ttc:.2f} €
""")
````

### Indexation et slicing

⚠️ **Piège classique pour les SQL-istes** : Python est **0-indexé**.
- Le 1er élément est à l'index **`0`**, pas `1`
- Le dernier élément est à l'index `len(liste) - 1`
- Ou plus simple : index **`-1`** pour le dernier, `-2` pour l'avant-dernier...

Le **slicing** permet d'extraire une **tranche** : `ma_liste[debut:fin]`
- `debut` est **inclus**, `fin` est **exclue** (comme `LIMIT/OFFSET` SQL : on prend `fin - debut` éléments)


````python
prix_produits = [19.90, 49.90, 12.50, 89.00, 34.90]
#                  0      1      2      3      4
#                 -5     -4     -3     -2     -1

# Accès par index
print(prix_produits[0])     # → 19.9  (premier)
print(prix_produits[2])     # → 12.5  (troisième)
print(prix_produits[-1])    # → 34.9  (dernier)
print(prix_produits[-2])    # → 89.0  (avant-dernier)

# Slicing : [debut:fin] — fin EXCLUE
print(prix_produits[1:3])   # → [49.9, 12.5]  (index 1 et 2, 3 est exclus)
# Slicing : [:fin] — depuis le début : fin EXCLUE
print(prix_produits[:2])    # → [19.9, 49.9]  (de 0 à 2, 2 est exclus on récupère bien 0 et 1)
# Slicing : [début:] — pas d'exclusion, va jusqu'à la fin
print(prix_produits[2:])    # → [12.5, 89.0, 34.9]  (de index 2 jusqu'à la fin)
print(prix_produits[-2:])   # → [89.0, 34.9]  (les 2 derniers, index -2 jusqu'à la fin)

# Modification d'un élément (les listes sont modifiables)
prix_produits[0] = 17.90
print(prix_produits)
````

### Méthodes et fonctions essentielles sur les listes

Deux familles à distinguer (et c'est inhabituel quand on vient de JS) :

- **Fonctions** (globales) : `len()`, `sum()`, `min()`, `max()`, `sorted()`
  → s'appellent sur la liste : `sum(ma_liste)`
- **Méthodes** (attachées à l'objet liste) : `.append()`, `.remove()`, `.sort()`
  → s'appellent après un point : `ma_liste.append(...)`

⚠️ Piège classique : `.sort()` **modifie la liste sur place** et ne retourne rien.
`sorted(ma_liste)` retourne une **nouvelle liste triée** sans toucher à l'originale.


````python
prix_produits = [19.90, 49.90, 12.50, 89.00, 34.90]

# Fonctions d'agrégation (les amies du data analyst — équivalents COUNT/SUM/MIN/MAX SQL)
print("Nombre :", len(prix_produits))        # COUNT(*)
print("Total :", sum(prix_produits))         # SUM(prix)
print("Min :", min(prix_produits))           # MIN(prix)
print("Max :", max(prix_produits))           # MAX(prix)
print("Moyenne :", sum(prix_produits) / len(prix_produits))   # AVG(prix)
print("Sorted :", sorted(prix_produits))           # liste triée
print("Liste intacte : ", prix_produits)

# Ajouter un élément à la fin
prix_produits.append(24.90)
print("Après append :", prix_produits)

# Retirer un élément par sa VALEUR (pas son index)
prix_produits.remove(12.50)
print("Après remove(12.50) :", prix_produits)

# Trier — deux variantes
copie_triee = sorted(prix_produits)            # nouvelle liste, l'originale est intacte
print("Triée (copie) :", copie_triee)
print("Originale :", prix_produits)

prix_produits.sort()                            # modifie SUR PLACE, ne retourne rien
print("Originale après .sort() :", prix_produits)

# Tester la présence d'une valeur
print(49.90 in prix_produits)    # True
print(99.99 in prix_produits)    # False

prix_produits.sort()
print("Méthode sort :", prix_produits)
````

### Les dictionnaires

Un **dictionnaire** stocke des paires **clé → valeur**.
C'est LA structure la plus importante de Python pour la data.

```python
produit = {"nom": "Casque", "prix": 49.90, "stock": 12}


````python
# Création d'un dictionnaire : accolades, paires clé: valeur
produit = {
    "nom": "Casque audio Bluetooth",
    "prix": 49.90,
    "stock": 12,
    "en_promo": True,
}

print(produit)
print(type(produit))

# Accès à une valeur par sa clé (entre crochets, comme un index — mais avec une clé)
print(produit["nom"])
print(produit["prix"])

# Modification d'une valeur existante
produit["prix"] = 39.90
print("Nouveau prix :", produit["prix"])

# Ajout d'une nouvelle clé (il suffit de l'affecter)
produit["categorie"] = "high-tech"
print(produit)

# Suppression d'une clé
del produit["en_promo"]
print(produit)
````

### Parcourir un dictionnaire : `.keys()`, `.values()`, `.items()`

Trois méthodes pour extraire ce qu'il y a dans un dict :

| Méthode      | Retourne                          | Cas d'usage                       |
|--------------|-----------------------------------|------------------------------------|
| `.keys()`    | les clés                          | "quelles colonnes ai-je ?"         |
| `.values()`  | les valeurs                       | calculs sur l'ensemble des valeurs |
| `.items()`   | les paires `(clé, valeur)`        | parcourir clé + valeur ensemble    |

Et l'opérateur **`in`** permet de tester si une **clé** existe dans le dict — équivalent du `WHERE colonne IS NOT NULL` SQL en quelque sorte.


````python
produit = {
    "nom": "Casque audio Bluetooth",
    "prix": 39.90,
    "stock": 12,
    "categorie": "high-tech",
}

# Les 3 méthodes essentielles
print(produit.keys())     # dict_keys([...])  → toutes les clés
print(produit.values())   # dict_values([...]) → toutes les valeurs
print(produit.items())    # dict_items([...]) → toutes les paires (clé, valeur)
list_keys = produit.keys()
print("list_keys : ", list_keys)
print(type(list_keys))

# On peut convertir en liste si on veut vraiment une liste
print(list(produit.keys()))
print(list(produit.values()))
list_values = (list(produit.values()))
print("list_values", list_values)
print(type(list_values))


# Tester la présence d'une CLÉ (pas d'une valeur) avec "in"
print("prix" in produit)          # True
print("date_promo" in produit)    # False
print("Casque audio Bluetooth" in produit)   # False ! (c'est une valeur, pas une clé)

# Accès sécurisé avec .get() — ne plante pas si la clé n'existe pas
print(produit.get("prix"))                     # 39.9
print(produit.get("date_promo"))               # None (pas d'erreur)
print(produit.get("date_promo", "non défini")) # "non défini" (valeur par défaut)

# À l'inverse, accès direct sur une clé inexistante → KeyError
# print(produit["date_promo"])   # Décommente pour voir l'erreur
````

C'est **LE concept clé** de cette section. On y mélange tout ce qu'on

```python
produits = [
    {"id": 1, "nom": "Casque",  "prix": 49.90, "stock": 12},

````python

# Notre "table produits" — une liste de dictionnaires
produits = [
    {"id": 1, "nom": "Casque",   "prix": 49.90, "stock": 12},
    {"id": 2, "nom": "Souris",   "prix": 19.90, "stock": 45},
    {"id": 3, "nom": "Clavier",  "prix": 79.00, "stock": 8},
    {"id": 4, "nom": "Écran",    "prix": 199.00, "stock": 3},
    {"id": 5, "nom": "Webcam",   "prix": 34.90, "stock": 27},
]

# COUNT(*) → combien de produits ?
print("Nombre de produits :", len(produits))

# Accès à UNE ligne entière (le premier produit)
print("Première ligne :", produits[0])

# Accès à UNE cellule (prix du 1er produit) : ligne d'abord, puis colonne
print("Prix du 1er produit :", produits[0]["prix"])

# Accès au NOM du dernier produit
print("Dernier produit :", produits[-1]["nom"])

# Modifier une "cellule" (UPDATE en SQL)
produits[0]["prix"] = 39.90
print("Après UPDATE :", produits[0])

# Ajouter une "ligne" (INSERT en SQL)
produits.append({"id": 6, "nom": "Tapis", "prix": 9.90, "stock": 50})
print("Après INSERT, nb produits :", len(produits))
````

### 🛠️ Mini-TP 3.3 — Statistiques sur un catalogue produits

Tu disposes d'une "table produits" e-commerce (sous forme de **liste de dicts**)

```python
produits = [
    {"id": 1, "nom": "Casque",   "prix": 49.90,  "stock": 12},
    {"id": 2, "nom": "Souris",   "prix": 19.90,  "stock": 45},
    {"id": 3, "nom": "Clavier",  "prix": 79.00,  "stock": 8},
    {"id": 4, "nom": "Écran",    "prix": 199.00, "stock": 3},
    {"id": 5, "nom": "Webcam",   "prix": 34.90,  "stock": 27},
]

prix_tous  = [49.90, 19.90, 79.00, 199.00, 34.90]
stock_tous = [12, 45, 8, 3, 27]
```

**Objectif** : produire ce rapport (2 décimales pour les montants) :

```
Catalogue : 5 produits
Stock total : 95 unités
Prix moyen : 76.54 €
Prix le plus cher : 199.00 €
Prix le moins cher : 19.90 €
Produit le plus cher : Écran
```

**Contraintes** :
- N'utilise QUE ce qu'on a vu : `len()`, `sum()`, `min()`, `max()`, indexation `[...]`, `.2f`
- Une **f-string** par ligne d'affichage
- Pour la dernière ligne (produit le plus cher), petite astuce :
  `prix_tous.index(max(prix_tous))` te renvoie l'**index** du produit le plus cher
  dans la liste — c'est une méthode des listes (`.index(valeur)` retourne la position
  d'une valeur). Ensuite, sers-toi de cet index pour récupérer le bon dict dans `produits`.


````python
produits = [
    {"id": 1, "nom": "Casque",   "prix": 49.90,  "stock": 12},
    {"id": 2, "nom": "Souris",   "prix": 19.90,  "stock": 45},
    {"id": 3, "nom": "Clavier",  "prix": 79.00,  "stock": 8},
    {"id": 4, "nom": "Écran",    "prix": 199.00, "stock": 3},
    {"id": 5, "nom": "Webcam",   "prix": 34.90,  "stock": 27},
]

prix_tous  = [49.90, 19.90, 79.00, 199.00, 34.90]
stock_tous = [12, 45, 8, 3, 27]

nb_produits = len(produits)
total_stock = sum(stock_tous)
prix_moyen = sum(prix_tous) / len(prix_tous)
prix_max = max(prix_tous)
prix_min = min(prix_tous)
index_max = prix_tous.index(prix_max)
nom_produit_max = produits[index_max]["nom"] 

print(f"Catalogue : {nb_produits} produits")
print(f"Stock total : {total_stock} unités")
print(f"Prix moyen : {prix_moyen:.2f} €")
print(f"Prix le plus cher : {prix_max:.2f} €")
print(f"Prix le moins cher : {prix_min:.2f} €")
print(f"Produit le plus cher : {nom_produit_max}")
````

### 3.4.1 L'indentation : la syntaxe qui structure

⚠️ **Spécificité Python n°1 — à connaître avant tout le reste.**

En Python, l'**indentation** (les espaces en début de ligne) n'est **pas cosmétique** :
c'est elle qui **définit les blocs de code**.

- En SQL, en Excel/VBA : tu peux tout coller sur une ligne, l'aération n'a aucun effet sémantique.
- En Python : **chaque niveau d'indentation = un bloc logique**. Il n'y a ni accolades `{ }`, ni `BEGIN/END`.

**Conventions** :
- **4 espaces** par niveau (pas de tabulation — VS Code convertit automatiquement)
- VS Code et Jupyter **auto-indentent** dès que tu termines une ligne par `:`
- L'analogie qui parle : c'est comme les **sous-niveaux d'un sommaire Word** :
  - `1. Titre` en marge
  - `    1.1 Sous-titre` décalé d'un cran
  - `        1.1.1 Sous-sous-titre` décalé de deux crans

Concrètement, dans un `if` :

```python
if condition:
    instruction_du_bloc_if    # ← indenté = dans le bloc du if
instruction_hors_du_bloc      # ← retour à la marge = en dehors du if
```

Si tu te trompes : **`IndentationError`** au lancement. Le message Python te pointe la ligne fautive — pas de panique.

### 3.4.2 Les conditions : if / elif / else

L'équivalent direct du `CASE WHEN ... THEN ... ELSE ... END` SQL.

```python
if condition_1:
    bloc_si_condition_1_vraie
elif condition_2:
    bloc_si_condition_2_vraie
else:
    bloc_sinon
```

- `if` est testé en premier
- `elif` (= "else if", mot-clé spécifique Python) n'est testé **que si** le `if` au-dessus est faux
- `else` est le filet de sécurité, exécuté si **toutes** les conditions précédentes ont été fausses
- `elif` et `else` sont optionnels — un `if` peut vivre seul

**Opérateurs de comparaison** (renvoient `True` ou `False`) :

| Opérateur  | Sens                          | Équivalent SQL    |
|------------|--------------------------------|-------------------|
| `==`       | égal                           | `=`               |
| `!=`       | différent                      | `<>` ou `!=`      |
| `<`, `>`   | strictement inférieur / supérieur | idem           |
| `<=`, `>=` | inférieur / supérieur ou égal  | idem              |

**Opérateurs booléens** :

| Python | Sens | SQL   |
|--------|------|-------|
| `and`  | et   | `AND` |
| `or`   | ou   | `OR`  |
| `not`  | non  | `NOT` |

⚠️ **Piège n°1 du débutant** : confondre `=` (assignation) et `==` (comparaison).
En SQL, `=` fait les deux selon le contexte. En Python, c'est strict :
- `prix = 30` → **assigne** 30 à `prix`
- `prix == 30` → **teste** si `prix` vaut 30 (renvoie `True` ou `False`)

Heureusement, Python te sauve : `if prix = 30:` provoque une `SyntaxError` immédiate. Mais le réflexe doit être ancré.

💡 **Rappel** (vu en 3.2 / 3.3) : les **valeurs falsy** (`0`, `0.0`, `""`, `[]`, `{}`, `None`, `False`) sont évaluées comme `False` dans un `if`. On s'en servira juste après.


````python
# Tarification d'un produit selon son prix : équivalent d'un CASE WHEN SQL
prix = 49.90

if prix > 100:
    categorie = "premium"
elif prix > 30:
    categorie = "standard"
else:
    categorie = "entrée de gamme"

print(f"Prix {prix:.2f} € → catégorie : {categorie}")

# Avec un autre prix pour activer une autre branche
prix = 12.50
if prix > 100:
    categorie = "premium"
elif prix > 30:
    categorie = "standard"
else:
    categorie = "entrée de gamme"
print(f"Prix {prix:.2f} € → catégorie : {categorie}")

# Combinaison de conditions avec and / or
prix = 49.90
stock = 12
en_promo = True

if prix > 30 and stock > 0:
    print("Produit éligible à la mise en avant")

if prix < 20 or en_promo:
    print("Produit en mode 'bonne affaire'")

# Tests sur des valeurs falsy — pattern data courant
panier = []
if not panier:
    print("Panier vide → on n'envoie pas de mail de confirmation")

description = ""
if not description:
    print("⚠️ Description produit manquante")
````

### 3.4.3 La boucle `for` : parcourir une collection

En Python, **`for` est un foreach par défaut**. C'est différent de ce que tu peux connaître en VBA ou Excel, où on écrit :

```vba
For i = 1 To 10
    ...
Next i
```

En Python, on n'a **pas besoin d'un compteur** : on parcourt **directement les éléments** d'une collection.

```python
for produit in produits:
    print(produit["nom"])
```

→ Se lit : *"pour chaque `produit` dans `produits`, faire …"*

**Analogie SQL** : c'est l'équivalent du `FOR EACH ROW` d'un **curseur** SQL — mais en bien plus simple. À chaque tour de boucle, la variable (`produit`) prend la valeur d'un élément de la collection, et le bloc indenté en dessous s'exécute.

**Convention de nommage** : variable de boucle au **singulier**, collection au **pluriel**. Ça se relit naturellement.

```python
for produit in produits:      # ✅ "pour chaque produit dans produits"
for prix in prix_tous:        # ✅ "pour chaque prix dans prix_tous"
for c in collection:          # ⚠️ correct mais cryptique
```

Si tu as **vraiment** besoin d'un index (numéro de ligne), on verra `range()` et `enumerate()` juste après.



````python
# Notre fil rouge : la liste de produits (réutilisée depuis le TP 3.3)
produits = [
    {"id": 1, "nom": "Casque",   "prix": 49.90,  "stock": 12},
    {"id": 2, "nom": "Souris",   "prix": 19.90,  "stock": 45},
    {"id": 3, "nom": "Clavier",  "prix": 79.00,  "stock": 8},
    {"id": 4, "nom": "Écran",    "prix": 199.00, "stock": 3},
    {"id": 5, "nom": "Webcam",   "prix": 34.90,  "stock": 27},
]

# Parcours simple : on affiche chaque produit (nom + prix)
for produit in produits:
    print(f"{produit['nom']} — {produit['prix']:.2f} €")

print()  # ligne vide pour aérer

# On peut boucler sur n'importe quel itérable, pas seulement une liste de dicts
prix_tous = [49.90, 19.90, 79.00, 199.00, 34.90]
for prix in prix_tous:
    print(f"Prix : {prix:.2f} €")

print()

# Même une chaîne est itérable (caractère par caractère)
for lettre in "DATA":
    print(lettre)
````

### `range()` — quand tu as besoin d'un index

La plupart du temps en data, on parcourt **directement les éléments** (`for produit in produits`). Mais parfois, on a besoin du **numéro de ligne** (l'index) — par exemple pour afficher "Produit n°1, n°2…" ou pour comparer la ligne courante à une autre.

`range(n)` génère une séquence d'**entiers de `0` à `n-1`** (cohérent avec l'indexation 0-based des listes).

```python
for i in range(5):
    print(i)    # → 0, 1, 2, 3, 4
```

**Variantes** :
- `range(stop)` → `0, 1, …, stop-1`
- `range(start, stop)` → commence à `start`
- `range(start, stop, step)` → avec un pas

Combiné à `len()` : `range(len(produits))` génère les index 0, 1, 2… de tous les produits.

🎯 Mais en pratique, on utilise plutôt **`enumerate()`** qui donne **l'index ET l'élément en même temps** — plus pythonique, plus lisible. À retenir comme le geste par défaut quand on veut "parcourir + connaître la position".


````python
# range(n) : entiers de 0 à n-1
for i in range(5):
    print(i)

print("---")

# Avec start, stop, step (de 2 à 10 par pas de 2)
for i in range(2, 10, 2):
    print(i)

print("---")

# Pattern "range(len(...))" : on récupère l'index, puis on accède à l'élément
for i in range(len(produits)):
    print(f"Produit n°{i + 1} : {produits[i]['nom']}")

print("---")

# Version pythonique avec enumerate() : index ET élément déballés en un coup
# (rappel 3.3 : on déballe le tuple (index, valeur) que renvoie enumerate)
for i, produit in enumerate(produits):
    print(f"Produit n°{i + 1} : {produit['nom']}")

print("---")

# enumerate(..., start=1) si on veut commencer la numérotation à 1
for num, produit in enumerate(produits, start=1):
    print(f"Ligne {num} : {produit['nom']} — {produit['prix']:.2f} €")
````

### Parcourir un dictionnaire avec `for`

On a vu en 3.3 que `.keys()`, `.values()` et `.items()` extraient des morceaux d'un dict. Avec un `for`, ces méthodes deviennent **vraiment** utiles.

| Méthode      | Ce qu'on récupère à chaque tour | Quand l'utiliser                                |
|--------------|----------------------------------|--------------------------------------------------|
| `.keys()`    | la **clé**                       | tu n'as besoin que des noms de colonnes          |
| `.values()`  | la **valeur**                    | tu veux opérer sur les valeurs uniquement        |
| `.items()`   | un tuple `(clé, valeur)`         | tu veux **les deux** — c'est le cas le plus fréquent |

À noter : par défaut, `for x in mon_dict:` boucle sur **les clés** (équivalent à `for x in mon_dict.keys()`). Mais en pratique, mieux vaut écrire `.keys()`, `.values()` ou `.items()` explicitement — ça enlève toute ambiguïté à la relecture.

🎯 **Le pattern à retenir** : `for cle, valeur in mon_dict.items():` — c'est celui qui apparaîtra dans 90% des codes Python data que tu liras.


````python
# Un seul produit (un dict)
produit = {
    "id": 1,
    "nom": "Casque audio Bluetooth",
    "prix": 49.90,
    "stock": 12,
}

# 1) Parcours des CLÉS uniquement
print("--- Clés ---")
for cle in produit.keys():
    print(cle)

# 2) Parcours des VALEURS uniquement
print("--- Valeurs ---")
for valeur in produit.values():
    print(valeur)

# 3) Parcours des PAIRES (clé, valeur) — le plus courant
print("--- Paires ---")
for cle, valeur in produit.items():
    print(f"{cle} : {valeur}")

# 4) Raccourci : par défaut, for x in dict boucle sur les clés
print("--- Sans .keys() (équivalent mais moins explicite) ---")
for cle in produit:
    print(cle)
````

### 3.4.4 La boucle `while` — mention rapide

**Principe** : *tant que* la condition est vraie, on continue.

```python
while condition:

````python

# Exemple 1 : compteur qui décrémente
compteur = 5
while compteur > 0:
    print(f"Compteur : {compteur}")
    compteur -= 1       # CRUCIAL : sans cette ligne, boucle infinie
print("Décollage !")

print("---")

# Exemple 2 : accumuler tant qu'on est sous un seuil
prix_tous = [49.90, 19.90, 79.00, 199.00, 34.90]
total = 0
i = 0

while total < 100 and i < len(prix_tous):
    total += prix_tous[i]
    print(f"Ajout de {prix_tous[i]:.2f} € → total cumulé = {total:.2f} €")
    i += 1

print(f"Seuil franchi après {i} produit(s) — total : {total:.2f} €")
````

### 3.4.5 Reproduire SQL avec des boucles — le climax

Tu maîtrises SQL. Tu sais écrire :
- `SELECT prix FROM produits` → extraire une colonne
- `SELECT * FROM produits WHERE stock > 10` → filtrer des lignes
- `SELECT COUNT(*) FROM produits WHERE stock = 0` → compter avec condition
- `SELECT SUM(prix * stock) FROM produits` → agréger

**Voici comment on fait la même chose en Python pur**, avec uniquement les briques qu'on vient de voir : listes, dicts, `for`, `if`, et le pattern **accumulateur**.

| SQL                                          | Python                                          |
|----------------------------------------------|-------------------------------------------------|
| `SELECT prix FROM produits`                  | boucle + `.append()` dans une liste             |
| `SELECT * FROM produits WHERE stock > 10`    | boucle + `if` + `.append()`                     |
| `SELECT COUNT(*) WHERE stock = 0`            | boucle + `if` + compteur                        |
| `SELECT SUM(prix * stock) FROM produits`     | boucle + accumulateur                           |

🎯 **Pourquoi c'est important** :
- Tu construis ton **modèle mental** des opérations data en code procédural — c'est exactement ce que pandas fera sous le capot.
- Demain, ces 4 patterns deviendront des **one-liners** pandas (`df["prix"]`, `df[df["stock"] > 10]`, `(df["stock"] == 0).sum()`, `(df["prix"] * df["stock"]).sum()`), mais tu sauras **ce qui tourne en dessous**.
- Tu peux **toujours retomber** sur cette forme si pandas te perd, ou si tu travailles sur des données qui ne sont pas en DataFrame (un JSON d'API par exemple).

On reprend la liste `produits` enrichie d'un produit en rupture (Tapis, stock=0) pour pouvoir illustrer les 4 patterns.


````python
# Fil rouge enrichi : on ajoute un produit en rupture pour illustrer les patterns suivants
produits = [
    {"id": 1, "nom": "Casque",   "prix": 49.90,  "stock": 12},
    {"id": 2, "nom": "Souris",   "prix": 19.90,  "stock": 45},
    {"id": 3, "nom": "Clavier",  "prix": 79.00,  "stock": 8},
    {"id": 4, "nom": "Écran",    "prix": 199.00, "stock": 3},
    {"id": 5, "nom": "Webcam",   "prix": 34.90,  "stock": 27},
    {"id": 6, "nom": "Tapis",    "prix": 9.90,   "stock": 0},   # ← stock à 0
]

# =====================================================
# Pattern 1 — SELECT prix FROM produits
# =====================================================
# Objectif : extraire UNE colonne (tous les prix) dans une liste à part

prix_tous = []                              # accumulateur : liste vide au départ
for produit in produits:
    prix_tous.append(produit["prix"])       # à chaque tour, on ajoute le prix

print("Prix extraits :", prix_tous)
print(f"Nombre de prix : {len(prix_tous)}")
print(f"Somme : {sum(prix_tous):.2f} €")
print(f"Moyenne : {sum(prix_tous) / len(prix_tous):.2f} €")
````

````python
# =====================================================
# Pattern 2 — SELECT * FROM produits WHERE stock > 10
# =====================================================
# Objectif : construire une nouvelle liste contenant les LIGNES (dicts entiers)
# qui satisfont une condition.

produits_bien_stockes = []
for produit in produits:
    if produit["stock"] > 10:
        produits_bien_stockes.append(produit)

print(f"{len(produits_bien_stockes)} produit(s) avec stock > 10 :")
for p in produits_bien_stockes:
    print(f"  - {p['nom']} (stock : {p['stock']})")

print()

# Variante : SELECT nom FROM produits WHERE stock > 10
# (on combine projection ET filtre — l'opération data la plus fréquente)
noms_bien_stockes = []
for produit in produits:
    if produit["stock"] > 10:
        noms_bien_stockes.append(produit["nom"])

print("Noms uniquement :", noms_bien_stockes)
````

````python
# =====================================================
# Pattern 3 — SELECT COUNT(*) FROM produits WHERE stock = 0
# =====================================================
# Objectif : compter combien de produits satisfont une condition.
# Pattern accumulateur n°2 : un COMPTEUR (un entier qui s'incrémente).

nb_ruptures = 0
for produit in produits:
    if produit["stock"] == 0:
        nb_ruptures += 1


print(f"Produits en rupture : {nb_ruptures}")

# Variante : compter les produits premium (prix > 100 €)
nb_premium = 0
for produit in produits:
    if produit["prix"] > 100:
        nb_premium += 1

print(f"Produits premium (> 100 €) : {nb_premium}")

# Variante avec condition composite (rupture OU stock pléthorique)
nb_extreme = 0
for produit in produits:
    if produit["stock"] == 0 or produit["stock"] > 30:
        nb_extreme += 1

print(f"Produits en stock extrême (=0 ou >30) : {nb_extreme}")
````

````python
# =====================================================
# Pattern 4 — SELECT SUM(prix * stock) FROM produits
# =====================================================
# Objectif : agréger (somme) une grandeur, éventuellement calculée par ligne.
# Pattern accumulateur n°3 : un TOTAL (un nombre qui s'additionne).

# Variante simple : somme du stock total
stock_total = 0
for produit in produits:
    stock_total += produit["stock"]

print(f"Stock total (toutes lignes) : {stock_total} unités")

# Le classique : CA potentiel = somme de (prix * stock) sur chaque ligne
ca_total = 0
for produit in produits:
    ca_total += produit["prix"] * produit["stock"]

print(f"CA potentiel total : {ca_total:.2f} €")

# Avec condition : CA des produits premium uniquement (SUM WHERE prix > 100)
ca_premium = 0
for produit in produits:
    if produit["prix"] > 100:
        ca_premium += produit["prix"] * produit["stock"]

print(f"CA potentiel des produits premium : {ca_premium:.2f} €")
````

### 🎯 Récap : les 3 patterns accumulateurs

Tu viens d'apprendre **les 3 gestes fondamentaux** pour agréger des données avec une boucle. Ils se retrouvent partout en Python (et c'est ce que pandas fait sous le capot demain).

| Pattern         | Initialisation  | Opération dans la boucle              | Usage SQL                       |
|-----------------|------------------|---------------------------------------|---------------------------------|
| **Liste**       | `resultat = []`  | `resultat.append(valeur)`             | `SELECT col [WHERE ...]`        |
| **Compteur**    | `n = 0`          | `n += 1` (sous condition)             | `SELECT COUNT(*) [WHERE ...]`   |
| **Somme**       | `total = 0`      | `total += valeur` (calcul possible)   | `SELECT SUM(...) [WHERE ...]`   |

**La structure commune** :

```python
accumulateur = valeur_de_depart       # [] ou 0
for element in collection:
    if condition:                      # optionnel
        accumulateur = ...             # .append, += 1, += valeur
```

C'est ce schéma qu'on a déroulé 4 fois en 3.4.5. **Si tu maîtrises ces 3 gestes**, tu peux reproduire à peu près n'importe quelle requête SQL d'agrégation. Le reste = combinatoire (deux compteurs, un compteur + un total, accumulateur de dict pour grouper, etc.).

💡 **Petit défi mental** avant la suite : comment calculer un MIN ou un MAX avec un accumulateur ?
- Indice : ce n'est ni une liste ni un compteur — c'est un 4e pattern (*"je garde la meilleure valeur vue jusqu'ici"*).
- En pratique, on utilise les fonctions globales `min()` / `max()` (vues en 3.3) donc on n'écrit jamais ce 4e pattern à la main. Mais c'est bon de savoir qu'il existe.


````python
prix_tous = [49.90, 19.90, 79.00, 199.00, 34.90, 9.90]

# MAX
prix_max = prix_tous[0]                  # on part du 1er prix
for prix in prix_tous[1:]:               # on parcourt les SUIVANTS
    if prix > prix_max:
        prix_max = prix                  # on remplace si on trouve mieux

print(prix_max)   # 199.0

# MIN — il suffit d'inverser le comparateur
prix_min = prix_tous[0]
for prix in prix_tous[1:]:
    if prix < prix_min:
        prix_min = prix

print(prix_min)   # 9.9

# Les built-in de Python font exactement ce qu'on vient d'écrire — en 2 lignes au lieu de 8
print(f"MAX : {max(prix_tous)}")
print(f"MIN : {min(prix_tous)}")
````

### 3.5.1 Pourquoi écrire ses propres fonctions

Une **fonction** est une **recette nommée et réutilisable**. On lui donne un nom, on définit ce qu'elle prend en entrée et ce qu'elle renvoie en sortie, et on peut l'appeler autant de fois qu'on veut, depuis n'importe où.

**Analogie Excel** : c'est l'équivalent de créer une **formule réutilisable** à laquelle on donne un nom métier, plutôt que de copier-coller la même formule dans cinquante cellules. Si la règle de calcul change, on la modifie **à un seul endroit**.

**Analogie SQL** : c'est l'équivalent d'une **fonction définie côté base** (`CREATE FUNCTION`), ou plus simplement d'une expression métier qu'on retape à l'identique dans une dizaine de requêtes — et qu'on finit par regretter de ne pas avoir factorisée.

**Pourquoi c'est particulièrement utile en data** :
- Le même calcul revient dans plusieurs analyses (CA potentiel = `prix × stock`, prix TTC à partir d'un HT, catégorisation d'un produit…). On le nomme **une fois**, on le réutilise **partout**.
- Le code devient plus **lisible** : `ca = ca_potentiel(prix, stock)` se comprend instantanément, alors que `ca = prix * stock` se relit "qu'est-ce que ça fait, déjà ?".
- Si la règle change (par exemple : changement de seuil de catégorisation), on modifie la fonction **à un seul endroit** au lieu de tous les blocs où le calcul est copié.

Concrètement, en 3.4, on a calculé plusieurs fois `produit["prix"] * produit["stock"]` à différents endroits. Dans cette section, on va lui donner un nom — `ca_potentiel(prix, stock)` — et l'utiliser comme n'importe quelle autre fonction Python.


### 3.5.2 Anatomie d'une fonction : `def`, paramètres, `return`

La structure d'une fonction Python tient en 4 éléments :

```python
def nom_de_la_fonction(parametre_1, parametre_2):
    # corps de la fonction (indenté de 4 espaces)
    # ... calculs internes ...
    return valeur_a_renvoyer
```

| Élément                          | Rôle                                                      |
|----------------------------------|------------------------------------------------------------|
| `def`                            | mot-clé qui signale "je définis une fonction"              |
| `nom_de_la_fonction`             | le nom qu'on lui donne (convention : **`snake_case`**)     |
| `(parametre_1, parametre_2)`     | les **entrées** attendues, entre parenthèses               |
| `:` en fin de ligne              | déclenche le bloc indenté (même règle que `if`/`for`)      |
| `return valeur`                  | renvoie la **sortie** à celui qui a appelé la fonction     |

**Analogie SQL** : `ROUND(x, 2)` prend deux entrées (`x` et `2`) et renvoie une valeur (l'arrondi). Une fonction Python a exactement le même rôle. La différence : ici, **c'est toi qui définis** ce qu'elle calcule.

**Pour appeler** la fonction ensuite, on procède comme avec n'importe quelle fonction Python :

```python
resultat = nom_de_la_fonction(valeur_1, valeur_2)
```

Le résultat (la valeur que la fonction a renvoyée via `return`) est stocké dans une variable et peut être **réutilisé** ensuite : affiché, additionné, passé à une autre fonction, etc.

⚠️ **Piège n°1** : oublier le `return`. Dans ce cas, la fonction renvoie `None` (rien d'exploitable). On reviendra dessus en détail dans la section suivante — c'est la confusion la plus fréquente du débutant.


````python
# Définition de notre première fonction
def ca_potentiel(prix, stock):
    return prix * stock

# Appel de la fonction avec deux valeurs (positionnelles)
ca_casque = ca_potentiel(49.90, 12)
print(f"CA potentiel du Casque : {ca_casque:.2f} €")

# Appel sur un autre produit
ca_ecran = ca_potentiel(199.00, 3)
print(f"CA potentiel de l'Écran : {ca_ecran:.2f} €")

# On peut directement utiliser le résultat dans une expression
total = ca_potentiel(49.90, 12) + ca_potentiel(199.00, 3)
print(f"Total (Casque + Écran) : {total:.2f} €")

# Recopie de la liste produits (rappel du fil rouge)
produits = [
    {"id": 1, "nom": "Casque",   "prix": 49.90,  "stock": 12},
    {"id": 2, "nom": "Souris",   "prix": 19.90,  "stock": 45},
    {"id": 3, "nom": "Clavier",  "prix": 79.00,  "stock": 8},
    {"id": 4, "nom": "Écran",    "prix": 199.00, "stock": 3},
    {"id": 5, "nom": "Webcam",   "prix": 34.90,  "stock": 27},
    {"id": 6, "nom": "Tapis",    "prix": 9.90,   "stock": 0},
]

# Appel de la fonction dans une boucle for (rappel 3.4.3)
print()
for produit in produits:
    ca = ca_potentiel(produit["prix"], produit["stock"])
    print(f"{produit['nom']:<10} : {ca:>8.2f} €")
````

### 3.5.3 `return` vs `print` : la distinction clé

**C'est LA confusion la plus fréquente du débutant.** À assimiler avant tout le reste.

| Instruction         | Ce qu'elle fait                                       | Valeur réutilisable ?   |
|---------------------|--------------------------------------------------------|--------------------------|
| `print(x)`          | **affiche** `x` à l'écran (texte dans la console)      | ❌ Non — `print` renvoie `None` |
| `return x`          | **renvoie** `x` à l'appelant de la fonction            | ✅ Oui — `x` peut être stocké, réutilisé, recombiné |

Une fonction qui se contente de `print` **affiche** quelque chose dans la console, mais ne **rend rien d'exploitable** — elle renvoie implicitement `None`. Une fonction qui `return` une valeur permet de **stocker, additionner, comparer, reformater** cette valeur ensuite.

**Comparaison concrète** :

```python
# Version "print" : on voit le résultat à l'écran, mais on ne peut RIEN en faire ensuite
def afficher_ca(prix, stock):
    print(prix * stock)

# Version "return" : on récupère le résultat dans une variable
def ca_potentiel(prix, stock):
    return prix * stock
```

Avec la version `return`, on peut écrire :

```python
total = ca_potentiel(49.90, 12) + ca_potentiel(199.00, 3)   # ✅ fonctionne
```

Avec la version `print`, **c'est impossible** : `afficher_ca(...)` renvoie `None`, et `None + None` provoque une erreur.

🎯 **Règle métier à retenir** : en data, on veut **quasiment toujours `return`**. La fonction doit produire une **valeur exploitable**, qu'on choisira ensuite d'afficher ou non. Le `print` est rarement à mettre **dans** une fonction — il est à mettre **autour**, dans le code qui appelle la fonction.


````python
# Version 1 — fonction qui PRINT (à éviter pour les calculs)
def afficher_ca(prix, stock):
    print(prix * stock)        # affiche mais ne renvoie rien

# Version 2 — fonction qui RETURN (le bon réflexe data)
def ca_potentiel(prix, stock):
    return prix * stock        # renvoie une valeur exploitable

# Appels comparés
print("--- Appels directs ---")
afficher_ca(49.90, 12)        # affiche : 598.8
ca_potentiel(49.90, 12)       # ne s'affiche pas tout seul (le résultat est juste renvoyé)

# On stocke le résultat de chaque fonction dans une variable
print("\n--- Que renvoie chaque fonction ? ---")
resultat_1 = afficher_ca(49.90, 12)     # affiche 598.8, et resultat_1 vaut None
resultat_2 = ca_potentiel(49.90, 12)    # n'affiche rien, et resultat_2 vaut 598.8

print(f"resultat_1 (issu de print) : {resultat_1}")
print(f"resultat_2 (issu de return) : {resultat_2}")

# Combinaison : seul return permet de chaîner les calculs
print("\n--- Combinaison de plusieurs appels ---")
total_ok = ca_potentiel(49.90, 12) + ca_potentiel(199.00, 3)
print(f"Total avec return : {total_ok:.2f} €")     # fonctionne

# Et avec la version print ? Décommente pour voir l'erreur :
# total_ko = afficher_ca(49.90, 12) + afficher_ca(199.00, 3)
# → TypeError: unsupported operand type(s) for +: 'NoneType' and 'NoneType'
````

### 3.5.4 Arguments par défaut

Un paramètre peut avoir une **valeur par défaut**, qui sera utilisée si l'appelant ne la fournit pas. Très utile pour le **cas habituel** qu'on ne veut pas retaper à chaque appel.

```python
def nom_fonction(parametre_obligatoire, parametre_avec_defaut=valeur_par_defaut):
    ...
```

**Exemple métier** : en France, la TVA standard est à 20 %. Quand on calcule un prix TTC, on a quasi toujours cette valeur — mais on doit pouvoir la surcharger pour les **taux réduits** (5,5 % sur les livres et médicaments, 10 % sur la restauration et le transport).

**Analogie Excel** : c'est l'équivalent des **arguments optionnels** d'une formule Excel — par exemple `RECHERCHEV(...)` qui a un 4e argument optionnel `[valeur_proche]` qu'on peut omettre.

⚠️ **Règle syntaxique importante** : dans la définition de la fonction, les paramètres **avec valeur par défaut doivent venir APRÈS** ceux qui n'en ont pas.

```python
def ok(prix_ht, taux_tva=0.20):       # ✅ valide
    ...

def pas_ok(taux_tva=0.20, prix_ht):   # ❌ SyntaxError
    ...
```


````python
# Fonction avec un argument par défaut (TVA standard à 20 %)
def prix_ttc(prix_ht, taux_tva=0.20):
    return prix_ht * (1 + taux_tva)

# Appel sans préciser le taux : il prend la valeur par défaut (20 %)
ttc_standard = prix_ttc(100)
print(f"Prix TTC standard : {ttc_standard:.2f} €")

# Appel en précisant un taux réduit (livres / médicaments : 5,5 %)
ttc_reduit = prix_ttc(100, 0.055)
print(f"Prix TTC à 5,5 % : {ttc_reduit:.2f} €")

# Appel avec un taux intermédiaire (restauration / transport : 10 %)
ttc_intermediaire = prix_ttc(100, 0.10)
print(f"Prix TTC à 10 %  : {ttc_intermediaire:.2f} €")

# Utilisation dans une boucle sur la liste produits (TVA 20 % implicite)
print()
for produit in produits:
    ttc = prix_ttc(produit["prix"])
    print(f"{produit['nom']:<10} : {produit['prix']:>6.2f} € HT -> {ttc:>6.2f} € TTC")
````

### 3.5.5 Arguments nommés (mots-clés)

Jusqu'ici, on a passé les arguments **par position** : le 1er argument va dans le 1er paramètre, le 2e dans le 2e, etc. Python permet aussi de passer les arguments **par leur nom** — on parle alors d'**arguments nommés** (ou *keyword arguments*).

```python
prix_ttc(100, 0.055)                       # positionnel
prix_ttc(100, taux_tva=0.055)              # mixte
prix_ttc(prix_ht=100, taux_tva=0.055)      # tout nommé
```

**Avantages** :
- **Lisibilité** : `prix_ttc(100, taux_tva=0.055)` se lit mieux que `prix_ttc(100, 0.055)` — on voit immédiatement à quoi sert chaque valeur.
- **Robustesse** : si l'auteur de la fonction modifie l'ordre des paramètres, le code qui utilise les noms continue à fonctionner.
- **Clarté pour les fonctions à plusieurs paramètres optionnels** : on précise uniquement ceux qu'on veut, dans n'importe quel ordre.

**Règles à connaître** :
- Les arguments **positionnels** doivent venir **avant** les nommés dans l'appel : `f(1, taux=0.2)` ✅ mais pas `f(taux=0.2, 1)` ❌
- On ne peut pas donner deux fois la même valeur (positionnel + nommé sur le même paramètre)

🎯 **Pont pandas (demain)** : presque tous les appels pandas utilisent **massivement les arguments nommés** :

```python
df.sort_values(by="prix", ascending=False)
df.groupby(by="categorie").agg({"prix": "mean"})
df.merge(autre_df, on="id", how="left")
```

Tu vas en croiser **partout**. La syntaxe que tu apprends ici est exactement celle qui sera utilisée demain.


````python
# Fonction définie en 3.5.4
def prix_ttc(prix_ht, taux_tva=0.20):
    return prix_ht * (1 + taux_tva)

# 1) Appel POSITIONNEL : la position détermine quel paramètre reçoit quelle valeur
ttc_1 = prix_ttc(100, 0.055)
print(f"Positionnel        : {ttc_1:.2f} €")

# 2) Appel MIXTE : premier en positionnel, deuxième en nommé
ttc_2 = prix_ttc(100, taux_tva=0.055)
print(f"Mixte              : {ttc_2:.2f} €")

# 3) Appel TOUT NOMMÉ : tous les arguments sont explicites
ttc_3 = prix_ttc(prix_ht=100, taux_tva=0.055)
print(f"Tout nommé         : {ttc_3:.2f} €")

# 4) Avec tout nommé, on peut MÊME inverser l'ordre
ttc_4 = prix_ttc(taux_tva=0.055, prix_ht=100)
print(f"Tout nommé inversé : {ttc_4:.2f} €")

# --- Pourquoi c'est plus lisible : exemple avec plusieurs paramètres optionnels ---
def appliquer_remise(prix, taux_remise=0.0, remise_fidelite=0.0, montant_fixe=0.0):
    return prix * (1 - taux_remise) * (1 - remise_fidelite) - montant_fixe

# Sans arguments nommés : on doit deviner ce que chaque valeur représente
print()
print("Sans nommer :", appliquer_remise(100, 0.10, 0, 5))

# Avec arguments nommés : l'intention est immédiate
print("En nommant  :", appliquer_remise(100, taux_remise=0.10, montant_fixe=5))
````

### 3.5.6 La portée des variables (local / global)

**Règle simple à mémoriser** :
- Une variable créée **dans** une fonction n'existe **que dans cette fonction**. Une fois la fonction terminée, la variable disparaît.
- On fait **entrer** les données dans la fonction par les **paramètres**, et on récupère le résultat par le **`return`**. Pas autrement.

Visuellement, une fonction = une **boîte fermée** :

```
              ┌─────────────────────┐
paramètres →  │   corps             │  →  return (valeur)
(entrées)     │   variables locales │     (sortie)
              └─────────────────────┘
```

Deux conséquences importantes :

1. **Une variable interne ne déborde pas vers l'extérieur.** Si tu crées `total = 0` dans une fonction, ce `total` n'existe pas en dehors. Ton code principal peut très bien avoir une autre variable `total` ailleurs, sans interférence.

2. **La fonction ne dépend pas de l'extérieur.** On évite de lire des variables "qui traînent dans le notebook" depuis l'intérieur d'une fonction. C'est tentant, mais ça crée des fonctions **fragiles** qui plantent quand on les réutilise ailleurs. Le bon réflexe : ce dont la fonction a besoin → en paramètre.

**Bénéfice métier** : tes fonctions deviennent **autonomes** et **réutilisables**. Tu peux les copier-coller dans un autre notebook ou un autre script — elles continueront à fonctionner tant qu'on leur passe leurs paramètres.


````python
# Cas 1 : variable créée DANS une fonction → invisible dehors
def calcul_ttc(prix_ht):
    taux_tva = 0.20                    # variable LOCALE à la fonction
    return prix_ht * (1 + taux_tva)

resultat = calcul_ttc(100)
print(f"Résultat : {resultat:.2f} €")  # OK : on utilise la valeur RENVOYÉE

# Si on essaie d'accéder à taux_tva en dehors de la fonction, décommente :
# print(taux_tva)
# → NameError: name 'taux_tva' is not defined
# La variable a disparu à la fin de la fonction.

# Cas 2 : ce que tu passes à une fonction n'affecte PAS l'original (pour les types simples)
prix_original = 100

def ajouter_dix(prix):
    prix = prix + 10                   # modifie la copie locale, pas l'original
    return prix

nouveau = ajouter_dix(prix_original)
print(f"prix_original : {prix_original}")   # 100 (inchangé)
print(f"nouveau       : {nouveau}")          # 110

# Cas 3 : le BON RÉFLEXE — tout entre par les paramètres, tout sort par return
def ca_potentiel(prix, stock):
    return prix * stock

# Cette fonction ne dépend de RIEN d'extérieur. On peut la copier-coller
# dans n'importe quel autre notebook : elle fonctionnera.
ca = ca_potentiel(49.90, 12)
print(f"CA : {ca:.2f} €")
````

### 3.5.7 (Bonus) Renvoyer plusieurs valeurs

Une fonction peut renvoyer **plusieurs valeurs d'un coup**, en les séparant par des virgules après le `return`.

```python
def fonction(...):
    ...
    return valeur_1, valeur_2, valeur_3
```

Techniquement, Python emballe ces valeurs dans un **tuple** (rappel 3.3 : structure figée, parenthèses optionnelles). À l'appel, on a deux options :

```python
resultat = fonction(...)              # on récupère le tuple complet : (val1, val2, val3)
a, b, c = fonction(...)               # on dépaquète directement sur plusieurs variables
```

C'est le même mécanisme de **dépaquetage** que celui vu avec `enumerate()` et `.items()` en 3.4 — la mécanique est identique, elle s'applique simplement au `return` d'une fonction.

**Cas d'usage typique** : une fonction qui calcule à la fois plusieurs valeurs liées entre elles — le min et le max d'une série, le HT et la TVA à partir d'un TTC, le résultat d'un calcul et un indicateur de qualité associé.

🎯 **Tu rencontreras ce pattern en pandas** : par exemple, `df.shape` renvoie un tuple `(nb_lignes, nb_colonnes)` qu'on dépaquète souvent en `n_lignes, n_colonnes = df.shape`.


````python
# Fonction qui renvoie DEUX valeurs : le prix min et le prix max d'un catalogue
def bornes_prix(liste_produits):
    prix = []
    for produit in liste_produits:
        prix.append(produit["prix"])
    return min(prix), max(prix)        # renvoie un tuple (min, max)

# Appel avec DÉPAQUETAGE : on récupère les 2 valeurs en une ligne
pmin, pmax = bornes_prix(produits)
print(f"Prix min  : {pmin:.2f} €")
print(f"Prix max  : {pmax:.2f} €")
print(f"Amplitude : {pmax - pmin:.2f} €")

# On peut aussi récupérer le TUPLE entier sans le dépaqueter
print()
bornes = bornes_prix(produits)
print(f"Tuple complet : {bornes}")
print(f"Type          : {type(bornes)}")

# Autre exemple : décomposer un prix TTC en (HT, TVA)
def ttc_vers_ht_et_tva(prix_ttc, taux_tva=0.20):
    prix_ht = prix_ttc / (1 + taux_tva)
    montant_tva = prix_ttc - prix_ht
    return prix_ht, montant_tva

ht, tva = ttc_vers_ht_et_tva(120)
print(f"\nPour un TTC de 120,00 € → HT = {ht:.2f} € | TVA = {tva:.2f} €")
````

### 3.5.8 (Bonus) `lambda` : à savoir lire

Tu croiseras dans la documentation pandas et sur Stack Overflow une syntaxe condensée pour définir une fonction **très courte**, en **une seule ligne**, sans même lui donner de nom :

```python
lambda parametres: expression_a_renvoyer
```

C'est simplement une **fonction écrite de façon compacte**. Pas de `def`, pas de `return` explicite — l'expression placée après les `:` est ce qui est automatiquement renvoyé.

**Exemple** : la fonction `ca_potentiel(prix, stock)` qu'on a écrite plus haut peut s'écrire en une ligne :

```python
def ca_potentiel(prix, stock):
    return prix * stock
```

devient :

```python
ca_potentiel = lambda prix, stock: prix * stock
```

Les deux versions sont strictement équivalentes.

🎯 **Objectif pédagogique** :
- **Savoir LIRE** une `lambda` → **obligatoire**, tu en croiseras partout en pandas
- **Savoir ÉCRIRE** une `lambda` → **non obligatoire** au début. Continue à utiliser `def` pour tes propres fonctions, c'est plus lisible.

**Pourquoi `lambda` existe** : pour passer une **mini-fonction à une autre fonction**, sans avoir à la définir formellement. C'est le cas d'usage archétypique en pandas :

```python
df["prix_ttc"] = df["prix_ht"].apply(lambda x: x * 1.20)
```

Ici, on crée une "fonction jetable" qui multiplie par 1,20, juste pour la passer à `.apply()`. La définir avec un `def` aurait été plus verbeux pour une opération aussi simple.

**Limites à connaître** :
- Une `lambda` ne peut contenir **qu'une seule expression** — pas de bloc `if/else` étendu, pas de boucle, pas de plusieurs lignes
- Dès que la logique fait plus que 2-3 caractères de calcul, mieux vaut repasser à `def`


````python
# 1) Version classique avec def
def ca_potentiel(prix, stock):
    return prix * stock

# 2) Version équivalente avec lambda (pour comparer la syntaxe)
ca_potentiel_lambda = lambda prix, stock: prix * stock

# Appel : identique dans les deux cas
print("def    :", ca_potentiel(49.90, 12))
print("lambda :", ca_potentiel_lambda(49.90, 12))

# Autre exemple : appliquer une remise de 10 %
remise_10 = lambda prix: prix * 0.90
print(f"\nPrix après remise 10 % : {remise_10(49.90):.2f} €")

# --- Cas d'usage TYPIQUE : passer une lambda à une autre fonction ---
# (préfiguration directe de ce qu'on fera avec .apply() en pandas)

# sorted() accepte un argument nommé "key" qui est... une fonction !
# On lui passe une lambda pour trier produits par CA potentiel décroissant
produits_tries = sorted(produits, key=lambda p: p["prix"] * p["stock"], reverse=True)

print("\nProduits triés par CA potentiel décroissant :")
for produit in produits_tries:
    ca = produit["prix"] * produit["stock"]
    print(f"  {produit['nom']:<10} : {ca:>7.2f} €")
````

### 📋 Récap 3.5 — Les fonctions

**Structure d'une fonction** :

```python
def nom_fonction(param_1, param_2=valeur_defaut):
    # corps indenté (4 espaces)
    # ... calculs ...
    return valeur
```

**Tableau récap** :

| Concept                  | À retenir                                                   |
|--------------------------|--------------------------------------------------------------|
| `def`                    | mot-clé qui déclare une fonction                             |
| Indentation du corps     | 4 espaces — même règle que `if`/`for`/`while` (cf. 3.4.1)    |
| `return`                 | renvoie une valeur exploitable par l'appelant                |
| Sans `return`            | la fonction renvoie `None` (rien d'exploitable)              |
| `print` vs `return`      | `print` affiche, `return` rend une valeur réutilisable       |
| Argument par défaut      | `param=valeur` dans la signature — paramètre devient optionnel |
| Argument nommé           | `f(param=valeur)` à l'appel — plus lisible                   |
| Portée                   | les variables créées dans la fonction restent locales        |
| Retour multiple          | `return a, b` renvoie un tuple, dépaquetable en `x, y = f()` |
| `lambda` (bonus)         | fonction compacte sur une ligne — à savoir lire              |

**Les 3 réflexes à garder** :

1. **Une fonction = une boîte fermée** — les entrées passent par les paramètres, les sorties par `return`. Rien d'extérieur ne traverse les parois.
2. **`return`, pas `print`** — la fonction doit produire une valeur exploitable. L'affichage est le rôle du code qui l'appelle.
3. **Nomme tes arguments** dès qu'il y a 3 paramètres ou plus — `f(prix=10, taux=0.2)` se relit mieux que `f(10, 0.2)`.

---

🎯 **Pont vers pandas (demain)**

Aujourd'hui, on appelle nos fonctions **une à une dans une boucle `for`** sur chaque ligne du catalogue :

```python
for produit in produits:
    ca = ca_potentiel(produit["prix"], produit["stock"])
    print(f"{produit['nom']} : {ca:.2f} €")
```

Demain avec pandas, on confie directement la fonction à la colonne, et pandas l'applique à chaque ligne **automatiquement** :

```python
df["ca"] = df.apply(lambda row: ca_potentiel(row["prix"], row["stock"]), axis=1)
```

Plus de boucle visible — pandas s'en charge pour toi. Mais le **modèle mental** reste exactement le même : une fonction qui prend des entrées, et qui renvoie une sortie.


### 🛠️ Mini-TP 3.5 — Rapport produit avec fonctions réutilisables

Tu disposes de la liste `produits` (6 produits, fil rouge du chapitre 3).

**Objectif** : produire un rapport ligne par ligne, où chaque produit est affiché avec son **nom**, sa **catégorie** et son **CA potentiel**.

### 3.6.2 Le motif universel : `connect → cursor → execute → fetch → close`

Chaque interaction Python ↔ BDD SQL suit toujours **les mêmes 5 étapes** — c'est le motif que vous allez répéter à chaque cellule de ce chapitre, et qui vous suivra jusqu'à Snowflake demain.

```
┌──────────────────────────────────────────────────┐
│  1. IMPORT       import sqlite3                  │
│  2. CONNECT      conn = sqlite3.connect(...)     │
│  3. CURSOR       cur = conn.cursor()             │
│  4. EXECUTE      cur.execute("SELECT ...")       │
│  5. FETCH        cur.fetchall()                  │
│     CLOSE        conn.close()                    │
└──────────────────────────────────────────────────┘
```

| Étape       | Rôle                                                                            |
|-------------|----------------------------------------------------------------------------------|
| **import**  | charge le module SQL                                                            |
| **connect** | ouvre une connexion vers la base (un fichier pour SQLite, un serveur pour Snowflake) |
| **cursor**  | crée un "exécuteur" — l'objet qui transmet les requêtes et garde le résultat    |
| **execute** | envoie une requête SQL à la base                                                |
| **fetch**   | récupère le résultat (sous forme de liste de tuples)                            |
| **close**   | ferme la connexion proprement                                                   |

**Analogie pratique** : c'est l'équivalent du parcours que vous faites dans un client SQL comme DBeaver ou DataGrip.

| Dans DBeaver / DataGrip          | En Python                  |
|-----------------------------------|----------------------------|
| Se connecter au serveur          | `sqlite3.connect(...)`     |
| Ouvrir une fenêtre de requête    | `conn.cursor()`            |
| Appuyer sur F5                   | `cur.execute(...)`         |
| Lire les résultats dans la grille | `cur.fetchall()`           |
| Quitter le client                | `conn.close()`             |

⚠️ **Spécificité importante de SQLite** : la connexion se fait vers un **fichier**, pas vers un serveur. La toute première fois qu'on appelle `sqlite3.connect("catalogue.db")` sur un fichier inexistant, **SQLite crée le fichier**. Pas d'erreur, pas de message, pas de prompt — il apparaît silencieusement dans votre dossier de travail.


````python
# Étape 1 — Import du module sqlite3 (livré avec Python, rien à installer)
import sqlite3

# Étape 2 — Connexion à la base
# Si le fichier "catalogue.db" n'existe pas encore, SQLite le crée silencieusement
conn = sqlite3.connect("catalogue.db")
print("Connexion ouverte :", conn)

# Étape 3 — Création du curseur (l'objet qui exécute les requêtes)
cur = conn.cursor()
print("Curseur créé      :", cur)

# Étape 4 — Exécution d'une requête de test
# sqlite_version() est une fonction SQL native : pas besoin de table pour la tester
cur.execute("SELECT sqlite_version()")

# Étape 5 — Récupération du résultat
resultat = cur.fetchall()
print("Résultat brut     :", resultat)

# Le résultat est une LISTE de TUPLES — un tuple par ligne retournée
# Ici la requête a renvoyé une seule ligne avec une seule colonne
# Donc : liste de 1 tuple contenant 1 valeur → resultat[0][0]
version = resultat[0][0]
print(f"Version SQLite    : {version}")

# Étape 6 — Fermeture de la connexion (toujours, à la fin)
conn.close()
print("Connexion fermée.")
````

### 3.6.3 Créer une table et insérer les données

On va maintenant **donner vie** à notre base : créer la table `produits`, y insérer les 6 lignes du fil rouge, puis vérifier le résultat.

Les requêtes SQL utilisées dans cette cellule, vous les connaissez déjà :

- `DROP TABLE IF EXISTS` — supprime la table si elle existe (utile pour rejouer la cellule sans erreur de doublon)
- `CREATE TABLE` — crée la structure (colonnes + types)
- `INSERT INTO ... VALUES (...)` — insère une ligne

Ce qui est **nouveau côté Python** :

#### 1. La méthode `executemany()`

C'est une variante de `execute()` qui **répète la même requête pour plusieurs lots de paramètres**. Beaucoup plus rapide et plus lisible que `execute()` dans une boucle.

```python
cur.executemany("INSERT INTO produits (...) VALUES (?, ?, ?, ?)", liste_de_tuples)
```

#### 2. ⚠️ `conn.commit()` — le piège n°1 du chapitre

Pour qu'un `INSERT`, `UPDATE` ou `DELETE` soit **réellement enregistré** dans le fichier, il faut appeler `conn.commit()` après. **Sans commit, les modifications sont silencieusement perdues** à la fermeture de la connexion.

C'est exactement la philosophie d'une **transaction SQL** que vous connaissez : tant que vous n'avez pas validé (`COMMIT`), rien n'est gravé dans le marbre. Côté Python, on ne tape pas `COMMIT` en SQL — on appelle la méthode `conn.commit()`.

**Note importante** : pour les `SELECT`, **pas besoin** de `commit()` puisqu'on ne modifie rien. Le commit ne concerne que les opérations qui changent les données (INSERT / UPDATE / DELETE) ou la structure (CREATE / DROP / ALTER).


````python
import sqlite3

# Étape 1 — Connexion (recrée le fichier s'il a été supprimé)
conn = sqlite3.connect("catalogue.db")
cur = conn.cursor()

# Étape 2 — DROP si la table existe déjà
# Permet de rejouer cette cellule sans erreur en cas de second passage
cur.execute("DROP TABLE IF EXISTS produits")
print("Table existante supprimée (si elle existait).")

# Étape 3 — CREATE TABLE
# 4 colonnes avec leurs types SQL (équivalents à ce qu'on aura en Snowflake, Postgres...)
cur.execute("""
    CREATE TABLE produits (
        id      INTEGER PRIMARY KEY,
        nom     TEXT    NOT NULL,
        prix    REAL    NOT NULL,
        stock   INTEGER NOT NULL
    )
""")
print("Table 'produits' créée.")

# Étape 4 — INSERT des 6 lignes du fil rouge
# Chaque tuple = une ligne à insérer (mêmes valeurs qu'en 3.4)
donnees = [
    (1, "Casque",   49.90, 12),
    (2, "Souris",   19.90, 45),
    (3, "Clavier",  79.00,  8),
    (4, "Écran",   199.00,  3),
    (5, "Webcam",   34.90, 27),
    (6, "Tapis",     9.90,  0),
]

# executemany() : on envoie la même requête 6 fois, avec 6 jeux de valeurs
# Les 4 "?" sont les emplacements paramétrés (détail en 3.6.6)
cur.executemany(
    "INSERT INTO produits (id, nom, prix, stock) VALUES (?, ?, ?, ?)",
    donnees,
)
print(f"{len(donnees)} lignes insérées.")

# Étape 5 — COMMIT
# Sans cette ligne, les 6 INSERT seraient silencieusement perdus à la fermeture
conn.commit()
print("Modifications enregistrées (commit).")

# Étape 6 — Vérification rapide : on compte les lignes (se fait en 2 étapes)
cur.execute("SELECT COUNT(*) FROM produits") #renvoie l'objet sqlite3.Cursor
# puis on récupère chaque ligne (l'ensemble de la liste de tuples) avec fetchall()
list_produits = cur.fetchall() # → [(6,)]      ← liste contenant 1 tuple, fetchall() vide ensuite la lite
print(f"\nLa table 'produits' : {list_produits} .")
list_produits2 = cur.fetchall() # → [] une liste vide
print(f"\nLa table 'produits' : {list_produits2} .") 
first_produit = list_produits[0]  # → (6,)        ← le premier (et seul) tuple
print(f"\n1er tuple de La table 'produits' : {first_produit} lignes.")
nb_lignes = list_produits[0][0] # → 6           ← la première (et seule) valeur du tuple
print(f"\nLa table 'produits' contient {nb_lignes} lignes.")

# Étape 7 — Fermeture
conn.close()
print("Connexion fermée.")
````

### 3.6.4 Lire les données avec `SELECT` (le résultat = liste de tuples)

Maintenant qu'on a 6 lignes dans la table, on peut les lire. La requête `SELECT * FROM produits` vous est familière — ce qui change, c'est **ce que Python en fait**.

**Bonne nouvelle** : pour une lecture (`SELECT`), **pas besoin de `conn.commit()`**. On ne modifie rien, donc rien à valider.

Le motif reste le même : `execute` → `fetch`.

```python
cur.execute("SELECT * FROM produits")
lignes = cur.fetchall()
```

#### Format du résultat — c'est important

`fetchall()` renvoie une **liste de tuples**, où :

- chaque **tuple** = une **ligne SQL**
- chaque **élément du tuple** = une **colonne SQL** (dans l'ordre des colonnes du `SELECT`)

```python
[
  (1, 'Casque',  49.9, 12),    # 1ère ligne — un tuple à 4 colonnes
  (2, 'Souris',  19.9, 45),    # 2e ligne
  (3, 'Clavier', 79.0,  8),
  ...
]
```

C'est **exactement** la structure tuple vue en 3.3.3 — sauf que cette fois, les tuples viennent d'une base SQL au lieu d'être écrits à la main. Vous pouvez les parcourir avec un `for`, les indexer par position (`ligne[0]`, `ligne[1]`…), ou les **déballer** (`id, nom, prix, stock = ligne`).

#### ⚠️ Petit inconvénient

On **perd l'accès par nom de colonne**. Pour lire le prix de la 3e ligne, on doit écrire `lignes[2][2]` — il faut **mémoriser l'ordre des colonnes**, ce qui devient vite pénible.

C'est pour ça qu'on verra **`row_factory`** en 3.6.7, qui permet d'accéder par nom (`ligne["prix"]`). Et c'est aussi pour ça que **pandas existe** : un DataFrame, c'est exactement ça avec les noms de colonnes en bonus.


````python
import sqlite3

# Ouverture
conn = sqlite3.connect("catalogue.db")
cur = conn.cursor()

# --- SELECT * : on récupère toutes les colonnes, toutes les lignes ---
cur.execute("SELECT * FROM produits")
lignes = cur.fetchall()

print(f"Type de 'lignes' : {type(lignes)}")
print(f"Nombre de lignes : {len(lignes)}")
print()

# Affichage brut : on voit la structure "liste de tuples"
print("--- Résultat brut ---")
for ligne in lignes:
    print(ligne)
print()

# Parcours avec accès aux colonnes par INDEX (l'ordre vient du SELECT)
# id=ligne[0], nom=ligne[1], prix=ligne[2], stock=ligne[3]
print("--- Accès par index ---")
for ligne in lignes:
    print(f"id={ligne[0]} | nom={ligne[1]:<10} | prix={ligne[2]:>7.2f} € | stock={ligne[3]:>2}")
print()

# Variante plus lisible : DÉBALLAGE du tuple (rappel 3.3.3)
# Chaque tuple est déballé sur 4 variables nommées en une seule ligne
print("--- Accès via déballage de tuple ---")
for id_, nom, prix, stock in lignes:
    print(f"id={id_} | nom={nom:<10} | prix={prix:>7.2f} € | stock={stock:>2}")
print()

# SELECT de colonnes spécifiques : la structure des tuples S'ADAPTE au SELECT
cur.execute("SELECT nom, prix FROM produits")
deux_cols = cur.fetchall()
print("--- SELECT nom, prix (2 colonnes uniquement) ---")
print(deux_cols)
print(f"\nChaque tuple contient maintenant {len(deux_cols[0])} valeurs.")

# Fermeture
conn.close()
print("\nConnexion fermée.")
````

### 3.6.5 Filtrer et agréger : SQL fait le travail à votre place

**Voici le moment où vous allez vraiment apprécier d'avoir une BDD sous la main.**

Dans la section 3.4 (Mini-TP), pour calculer le CA potentiel par catégorie, on avait écrit en Python :

- une **boucle `for`** sur la liste des produits
- un **`if / elif / else`** pour catégoriser chaque produit
- **trois compteurs + trois accumulateurs CA** pour cumuler par catégorie
- un **bloc d'affichage** formaté

Soit **une bonne dizaine de lignes** de Python pur.

Avec SQL, le même calcul tient en **une seule requête**. Pourquoi ? Parce que **agréger des données est le métier d'origine de SQL** — c'est ce pour quoi le langage a été conçu il y a 50 ans. Python est plus généraliste : il sait tout faire, mais ce n'est pas l'outil le plus efficace pour des agrégats massifs.

#### La règle d'or à retenir

> **Tout ce qui s'exprime en SQL standard (`WHERE`, `GROUP BY`, `JOIN`, `ORDER BY`…) doit être fait en SQL.**
> Plus court, plus rapide, et déchargé sur le moteur de BDD (qui a été optimisé pendant des décennies pour ça).
>
> **Python prend le relais pour ce que SQL ne sait pas bien faire** : orchestration, logique métier complexe, appels d'API, machine learning, visualisation, automatisation.

Cette répartition des rôles, vous la garderez toute votre carrière en data analyse. C'est **l'un des messages essentiels** du chapitre.

#### Ce qu'on va voir dans la cellule de code

| Clause SQL                       | Équivalent Python en 3.4                  |
|-----------------------------------|--------------------------------------------|
| `WHERE`                           | boucle + `if` + `.append()`               |
| `ORDER BY`                        | `sorted(..., key=...)`                    |
| `COUNT`, `SUM`, `AVG`, `MIN`, `MAX` | accumulateurs Python                     |
| `CASE WHEN ... THEN ... END`      | `if / elif / else` à l'intérieur d'une boucle |
| `GROUP BY`                        | dictionnaire d'accumulateurs par clé      |

Tous ces patterns Python, vous les avez écrits **à la main** en 3.4. Vous allez voir leur équivalent SQL — c'est **bien plus court** et c'est ce que vous écrirez en pratique 95 % du temps.


````python
import sqlite3

conn = sqlite3.connect("catalogue.db")
cur = conn.cursor()

# =====================================================
# 1) WHERE — filtrer des lignes
# =====================================================
# Équivalent du "if produit['stock'] > 10: liste.append(...)" qu'on a écrit en 3.4
print("--- 1) WHERE : produits avec stock > 10 ---")
cur.execute("SELECT nom, stock FROM produits WHERE stock > 10")
for nom, stock in cur.fetchall():
    print(f"  {nom:<10} → stock = {stock}")
print()

# =====================================================
# 2) ORDER BY — trier les résultats
# =====================================================
# Équivalent du sorted(..., key=lambda p: p['prix'], reverse=True) vu en 3.5.8
print("--- 2) ORDER BY : produits triés par prix décroissant ---")
cur.execute("SELECT nom, prix FROM produits ORDER BY prix DESC")
for nom, prix in cur.fetchall():
    print(f"  {nom:<10} → {prix:>7.2f} €")
print()

# =====================================================
# 3) AGRÉGATIONS GLOBALES — COUNT, SUM, AVG, MIN, MAX
# =====================================================
# Équivalent de len() + sum() + (sum/len) + min() + max() en Python
print("--- 3) Agrégations globales (5 mesures en une requête) ---")
cur.execute("""
    SELECT
        COUNT(*),
        SUM(prix * stock),
        AVG(prix),
        MIN(prix),
        MAX(prix)
    FROM produits
""")
# fetchone() suffit : une seule ligne renvoyée par les agrégats globaux
nb, ca_total, prix_moyen, prix_min, prix_max = cur.fetchone()
print(f"  Nombre de produits : {nb}")
print(f"  CA total           : {ca_total:>8.2f} €")
print(f"  Prix moyen         : {prix_moyen:>8.2f} €")
print(f"  Prix min           : {prix_min:>8.2f} €")
print(f"  Prix max           : {prix_max:>8.2f} €")
print()

# =====================================================
# 4) CASE WHEN + GROUP BY — LE CLIMAX
# =====================================================
# Équivalent EXACT du Mini-TP 3.4 (10 lignes de Python), en UNE requête SQL
print("--- 4) CA potentiel par catégorie (Mini-TP 3.4 en 1 requête !) ---")
cur.execute("""
    SELECT
        CASE
            WHEN prix > 100 THEN 'Premium'
            WHEN prix > 30  THEN 'Standard'
            ELSE                 'Entrée de gamme'
        END AS categorie,
        COUNT(*)            AS nb_produits,
        SUM(prix * stock)   AS ca_potentiel
    FROM produits
    GROUP BY categorie
    ORDER BY ca_potentiel DESC
""")
for categorie, nb, ca in cur.fetchall():
    print(f"  {categorie:<16} | {nb} produit(s) | {ca:>9.2f} €")

conn.close()
````

### 3.6.6 Requêtes paramétrées : le `?`

Vous avez croisé les `?` en 3.6.3 dans la requête d'INSERT :

```python
cur.executemany("INSERT INTO produits (...) VALUES (?, ?, ?, ?)", donnees)
```

Le `?` est un **emplacement paramétré**. C'est la syntaxe à utiliser **dès qu'une valeur Python doit être insérée dans une requête SQL** — ce qui arrive constamment : filtrer par un id saisi, insérer des données issues d'un CSV, supprimer une ligne dynamique…

#### Le mauvais réflexe (à NE JAMAIS faire)

Tentation naturelle quand on débute : utiliser une **f-string** pour assembler la requête.

```python
# ❌ À NE JAMAIS FAIRE
nom_recherche = "Casque"
cur.execute(f"SELECT * FROM produits WHERE nom = '{nom_recherche}'")
```

Pourquoi c'est dangereux — deux raisons.

**1. Injection SQL** (risque de sécurité)

Si la variable vient d'un utilisateur, d'un formulaire ou d'un fichier non maîtrisé, elle peut contenir **du SQL malveillant** :

```python
nom_recherche = "Casque'; DROP TABLE produits; --"
cur.execute(f"SELECT * FROM produits WHERE nom = '{nom_recherche}'")
# → la requête finale devient :
# SELECT * FROM produits WHERE nom = 'Casque'; DROP TABLE produits; --'
# Résultat : la table est SUPPRIMÉE silencieusement
```

C'est **l'attaque la plus connue du Web** (référencée OWASP Top 10 depuis 20 ans). Même si dans ton notebook personnel ce n'est probablement pas une menace immédiate, **adopter le réflexe `?` dès maintenant** te protégera quand tu travailleras sur des données externes ou en production.

**2. Gestion automatique des apostrophes, types et dates**

Avec une f-string, tu dois gérer toi-même :
- L'échappement des apostrophes : `O'Connor` deviendrait `'O'Connor'` → SQL invalide
- Le formatage des nombres : `49.90` ou `49,90` selon la locale ?
- L'encodage des dates : `'2026-05-30'` ou `'30/05/2026'` ?

Le `?` paramétré **gère tout ça pour toi**, automatiquement et correctement.

#### Le bon réflexe

```python
nom_recherche = "Casque"
cur.execute("SELECT * FROM produits WHERE nom = ?", (nom_recherche,))
```

- La requête contient un `?` à la place de la valeur
- La valeur est passée en **deuxième argument**, sous forme de **tuple**
- SQLite remplace le `?` après avoir échappé et typé proprement la valeur

⚠️ **Détail subtil** : le deuxième argument est **toujours un tuple ou une liste** — même pour une seule valeur. Donc `(nom_recherche,)` avec **la virgule traînante** (rappel 3.3.3 — c'est elle qui en fait un tuple à un élément). Sans la virgule, `(nom_recherche)` n'est qu'une chaîne entre parenthèses, et sqlite3 lèvera une erreur.

#### Universalité

Cette syntaxe `?` est **standard SQL DB-API 2.0** — tu la retrouveras strictement à l'identique avec **snowflake-connector** demain, et aussi avec **psycopg2** (PostgreSQL), **mysql-connector**, etc. **Apprendre une fois, réutiliser partout.**


````python
import sqlite3

conn = sqlite3.connect("catalogue.db")
cur = conn.cursor()

# =====================================================
# 1) Requête paramétrée simple — UNE valeur
# =====================================================
nom_recherche = "Casque"

# Le ? est remplacé par la valeur du tuple (avec virgule traînante !)
cur.execute("SELECT * FROM produits WHERE nom = ?", (nom_recherche,))
print(f"--- Recherche par nom = '{nom_recherche}' ---")
for ligne in cur.fetchall():
    print(f"  {ligne}")
print()

# =====================================================
# 2) Plusieurs paramètres — l'ordre des ? = l'ordre du tuple
# =====================================================
prix_min = 30
prix_max = 100

cur.execute("""
    SELECT nom, prix
    FROM produits
    WHERE prix BETWEEN ? AND ?
    ORDER BY prix
""", (prix_min, prix_max))

print(f"--- Produits entre {prix_min} € et {prix_max} € ---")
for nom, prix in cur.fetchall():
    print(f"  {nom:<10} → {prix:>7.2f} €")
print()

# =====================================================
# 3) Filtre dynamique (style "saisie utilisateur")
# =====================================================
seuil_stock = 10

cur.execute("""
    SELECT nom, stock
    FROM produits
    WHERE stock > ?
    ORDER BY stock DESC
""", (seuil_stock,))

print(f"--- Produits avec stock > {seuil_stock} ---")
for nom, stock in cur.fetchall():
    print(f"  {nom:<10} → stock = {stock}")
print()

# =====================================================
# 4) Démonstration de sécurité : ce que la f-string AURAIT cassé
# =====================================================
# Saisie potentiellement malveillante (utilisateur, fichier non maîtrisé...)
nom_malveillant = "Casque'; DROP TABLE produits; --"

# Version PARAMÉTRÉE — SAFE
# Le ? traite la chaîne ENTIÈRE comme une valeur littérale, pas comme du SQL
cur.execute("SELECT * FROM produits WHERE nom = ?", (nom_malveillant,))
print("--- Recherche paramétrée avec valeur 'malveillante' ---")
resultat = cur.fetchall()
print(f"  Résultat : {resultat}")
print("  → liste vide (aucun produit n'a ce nom), mais surtout : la TABLE est INTACTE.")
print()

# Vérification : la table existe toujours et contient bien 6 lignes
cur.execute("SELECT COUNT(*) FROM produits")
print(f"  Nombre de produits dans la table : {cur.fetchone()[0]}")

conn.close()
````

### 3.6.7 Récupérer en dictionnaires (`row_factory`)

En 3.6.4, on a vu que `fetchall()` renvoie des **tuples** : on accède aux colonnes par **index** (`ligne[2]`). Inconvénient : il faut mémoriser l'ordre des colonnes, et le code casse si on modifie le `SELECT`.

**La solution** : configurer la connexion avec `conn.row_factory = sqlite3.Row`. Après cette ligne, chaque ligne renvoyée par `fetchall()` se comporte comme un **objet avec accès par nom** : `ligne["prix"]` au lieu de `ligne[2]`.

```python
conn = sqlite3.connect("catalogue.db")
conn.row_factory = sqlite3.Row       # ← à activer AVANT de créer le curseur
cur = conn.cursor()

cur.execute("SELECT nom, prix FROM produits")
for ligne in cur.fetchall():
    print(ligne["nom"], ligne["prix"])    # accès par NOM
```

#### Avantage immédiat

| Sans `row_factory`                       | Avec `row_factory = sqlite3.Row`            |
|------------------------------------------|---------------------------------------------|
| `ligne[2]`                               | `ligne["prix"]`                             |
| il faut mémoriser l'ordre des colonnes   | le nom métier est explicite                 |
| code fragile si on modifie le `SELECT`   | code robuste — le nom reste valide          |

L'objet `sqlite3.Row` est **hybride** :
- accès par **index** (`ligne[0]`) — toujours valide
- accès par **nom** (`ligne["nom"]`) — nouveau
- liste des noms de colonnes via `ligne.keys()`

---

#### ⚠️ Limite : `sqlite3.Row` n'est PAS un dict Python normal

C'est un objet **spécifique au module sqlite3**, en **lecture seule**. Pour 90 % des usages dans la cellule où tu fais le `SELECT`, ça suffit. Mais dès que tu veux faire l'une de ces 4 choses :

| Besoin                                                       | Pourquoi `sqlite3.Row` ne suffit pas        |
|--------------------------------------------------------------|----------------------------------------------|
| **Modifier** une valeur (changer un prix, ajouter une colonne) | la Row est en lecture seule                  |
| **Sérialiser** en JSON (envoi à une API, écriture dans un fichier) | la Row n'est pas reconnue par `json.dumps()` |
| **Passer la ligne à une autre bibliothèque** (pandas, requests…) | les libs attendent un `dict` standard        |
| **Afficher** lisiblement avec `print(ligne)`                  | la Row affiche `<sqlite3.Row at 0x...>`      |

…il faut la **convertir en vrai dict Python** avec `dict(ligne)`. C'est une opération simple, instantanée, qu'on fait dès qu'on quitte le périmètre "lecture immédiate dans la cellule".

#### Conversion en masse : reconstituer la structure du chapitre 3.3

Pour convertir **toutes les lignes** d'un coup en liste de dicts (la structure qu'on avait inventée à la main en 3.3) :

```python
produits_liste = [dict(ligne) for ligne in lignes]
```

C'est la **boucle d'or** : tu pars d'une BDD SQL, tu obtiens une liste de dicts Python prête à l'emploi.

---

#### Lien avec les alias `AS` de 3.6.6

Quand on accède par nom, les **alias `AS`** prennent tout leur sens. Si tu écris `SELECT COUNT(*) AS nb_produits`, tu y accèdes avec `ligne["nb_produits"]`. Sans alias, tu aurais `ligne["COUNT(*)"]` — illisible. **D'où le réflexe d'aliasser systématiquement les colonnes calculées.**

---

#### 🎯 La passerelle vers pandas

Cette philosophie — *"une ligne = un objet avec colonnes nommées"* — c'est **exactement** ce qu'est une **ligne de DataFrame pandas**. Quand vous écrirez demain `df = pd.read_sql_query("SELECT * FROM produits", conn)`, vous retrouverez la même mécanique, en plus puissant. `row_factory` est le **pas intermédiaire** qui prépare votre cerveau à pandas.


````python
import sqlite3

conn = sqlite3.connect("catalogue.db")

# ↓ La seule ligne qui change tout : on demande à sqlite3 de renvoyer
#   les lignes comme objets nommés, au lieu de tuples bruts.
# ⚠️ À placer AVANT la création du curseur, sinon ça ne s'applique pas.
conn.row_factory = sqlite3.Row

cur = conn.cursor()

# =====================================================
# ÉTAPE 1 — Récupérer les données avec accès par NOM
# =====================================================
cur.execute("SELECT * FROM produits")
lignes = cur.fetchall()

print(f"Type d'une ligne : {type(lignes[0])}")
print(f"Colonnes dispo   : {lignes[0].keys()}")
print()

# Accès par NOM (la nouveauté du row_factory)
print("--- Accès par nom de colonne ---")
for ligne in lignes:
    print(f"  id={ligne['id']} | nom={ligne['nom']:<10} | prix={ligne['prix']:>7.2f} € | stock={ligne['stock']:>2}")
    print(type(ligne)) # -> <class 'sqlite3.Row'> (en lecture seule, pas un dict)
print()

# Accès par INDEX (continue de marcher en parallèle — la Row est hybride)
print("--- Accès par index (toujours valide) ---")
for ligne in lignes:
    print(f"  ligne[0]={ligne[0]}, ligne[1]={ligne[1]}")
print()

# =====================================================
# ÉTAPE 2 — Limite : la Row est en LECTURE SEULE
# =====================================================
# Tu peux LIRE n'importe quelle colonne, mais tu ne peux pas la MODIFIER directement.
print("--- Limite : on ne peut PAS modifier une sqlite3.Row ---")
ligne = lignes[0]
print(f"  Lecture OK   : ligne['prix'] = {ligne['prix']}")

# Décommente la ligne suivante pour voir l'erreur :
# ligne["prix"] = 99.99
# → TypeError: 'sqlite3.Row' object does not support item assignment

print("  Modification : interdite (TypeError si on essaie)")
print("  → Pour modifier, il faut d'abord convertir en dict normal (étape 3).")
print()

# =====================================================
# ÉTAPE 3 — Conversion en dict Python normal
# =====================================================
# La fonction dict() transforme la Row "lecture seule" en dictionnaire modifiable.
print("--- Conversion en dict Python normal ---")
premier = dict(lignes[0])
print(f"  type    : {type(premier)}")
print(f"  contenu : {premier}")

# Maintenant on peut modifier librement
premier["prix"] = 39.90                 # change une valeur existante
premier["categorie"] = "Premium"        # ajoute une nouvelle clé
print(f"  modifié : {premier}")
print()

# =====================================================
# ÉTAPE 4 — Conversion en masse : liste de dicts
# =====================================================
# On reconstitue la structure du chapitre 3.3, mais alimentée par SQL.
produits_liste = [dict(ligne) for ligne in lignes]
print("--- Liste de dicts (structure 3.3 reconstituée depuis SQLite) ---")
for p in produits_liste:
    print(f"  {p}")
print()

# =====================================================
# ÉTAPE 5 — Cas d'usage concret : sérialisation JSON
# =====================================================
# Tu en auras besoin pour : envoyer à une API, écrire dans un fichier .json,
# transmettre à un collègue qui utilise un autre outil...
import json

# Décommente pour voir l'erreur avec une Row brute :
# json.dumps(lignes[0])
# → TypeError: Object of type Row is not JSON serializable

# La conversion en dict débloque tout
# (ensure_ascii=False préserve les accents type "Écran" au lieu de "\u00c9cran")
json_premier = json.dumps(dict(lignes[0]), ensure_ascii=False)
print("--- Conversion JSON (cas réel : envoi vers API ou fichier) ---")
print(f"  {json_premier}")
print()

# =====================================================
# ÉTAPE 6 — Bonus : combiner row_factory + alias AS
# (rappel 3.6.6 : les alias permettent d'utiliser des noms métier)
# =====================================================
cur.execute("""
    SELECT
        CASE
            WHEN prix > 100 THEN 'Premium'
            WHEN prix > 30  THEN 'Standard'
            ELSE 'Entrée de gamme'
        END                 AS categorie,
        COUNT(*)            AS nb_produits,
        SUM(prix * stock)   AS ca_potentiel
    FROM produits
    GROUP BY categorie
    ORDER BY ca_potentiel DESC
""")
print("--- Agrégation avec accès par alias AS ---")
for ligne in cur.fetchall():
    print(f"  {ligne['categorie']:<16} | {ligne['nb_produits']} produit(s) | {ligne['ca_potentiel']:>9.2f} €")

conn.close()
````

### 📋 3.6.8 Récap + pont vers pandas

Vous venez d'apprendre **le motif universel d'accès aux bases SQL depuis Python**. Tout ce qui suit dans votre carrière data — Snowflake, PostgreSQL, MySQL, Oracle… — utilisera exactement la même mécanique.

#### Le motif en 5 + 1 étapes

```
┌──────────────────────────────────────────────────┐
│  1. IMPORT       import sqlite3                  │
│  2. CONNECT      conn = sqlite3.connect(...)     │
│  3. CURSOR       cur = conn.cursor()             │
│  4. EXECUTE      cur.execute("SELECT ...", (...)) │
│  5. FETCH        cur.fetchall()  /  fetchone()   │
│     COMMIT       conn.commit()   ← uniquement si modif  │
│  6. CLOSE        conn.close()                    │
└──────────────────────────────────────────────────┘
```

#### Les briques essentielles vues dans le chapitre

| Brique                          | À retenir                                                          |
|---------------------------------|---------------------------------------------------------------------|
| `sqlite3.connect("fichier.db")` | crée le fichier s'il n'existe pas                                  |
| `cur.execute(sql)`              | lance la requête — **ne renvoie pas les données**                  |
| `cur.fetchall()`                | renvoie **liste de tuples** — **vide ensuite le curseur**          |
| `cur.fetchone()`                | renvoie **un seul tuple** — pratique quand on attend 1 ligne       |
| `conn.commit()`                 | **obligatoire** après INSERT / UPDATE / DELETE                     |
| `conn.close()`                  | habitude pro — toujours fermer ce qu'on a ouvert                   |
| `executemany(sql, liste)`       | insère plusieurs lignes en une seule opération                     |
| `?` paramétré                   | **toujours** pour les valeurs Python — jamais de f-string SQL      |
| `AS alias`                      | renomme une colonne — utile avec `row_factory` et pandas           |
| `conn.row_factory = sqlite3.Row` | accès aux colonnes par **nom** au lieu d'index                    |

#### Les 4 pièges à ne JAMAIS oublier

1. **`conn.commit()` après INSERT/UPDATE/DELETE** — sinon les modifications sont perdues silencieusement.
2. **Toujours `?` paramétré, jamais de f-string en SQL** — sécurité (injection SQL) + gestion automatique des apostrophes et types.
3. **`fetchall()` vide le curseur** — il faut le **stocker dans une variable** si on veut le réutiliser, ou refaire un `execute()`.
4. **Tuple à 1 élément = virgule traînante** — `(valeur,)` et pas `(valeur)`.

---

#### 🎯 Pont vers pandas (chapitre 3.7, juste après)

Maintenant que vous comprenez ce qu'il y a sous le capot, vous allez voir à quel point **pandas raccourcit le code**.

**Tout ce qu'on vient d'écrire en 5 lignes** :

```python
conn = sqlite3.connect("catalogue.db")
conn.row_factory = sqlite3.Row
cur = conn.cursor()
cur.execute("SELECT * FROM produits WHERE prix > 30")
lignes = cur.fetchall()
# puis boucle for, conversion en dict, etc.
```

**…devient une seule ligne en pandas** :

```python
import pandas as pd
df = pd.read_sql_query("SELECT * FROM produits WHERE prix > 30", conn)
```

Et `df` est directement un **DataFrame** : tableau navigable avec noms de colonnes, slicing, filtrage, agrégations, jointures, exports… le tout sans boucle `for`.

#### Ce que vous gagnez à avoir fait ce chapitre malgré tout

- **Vous comprenez ce que `read_sql_query` fait** sous le capot. Quand pandas se comporte de façon inattendue, vous savez où chercher.
- **Vous pouvez écrire du SQL pur quand pandas n'est pas approprié** — scripts d'INSERT/UPDATE en masse, jobs de chargement quotidiens, interventions ponctuelles sans charger une grosse dépendance.
- **Le motif `connect → cursor → execute → fetch → close` est universel**. Demain, en J2, vous le retrouverez tel quel avec Snowflake. C'est un investissement pour toute votre carrière.

> **Résumé en une phrase** : *pandas vous fait gagner du temps, sqlite3 vous fait gagner en compréhension. Vous avez besoin des deux.*


### 🛠️ Mini-TP 3.6 — Rapport "produits à réapprovisionner"

Vous travaillez au service supply chain de Cdiscount. On vous demande de produire un rapport listant **les produits dont le stock est descendu sous un certain seuil**, pour préparer un bon de commande.

**Données** : la table `produits` créée en 3.6.3 dans `catalogue.db` — elle est déjà en place, pas besoin de la recréer.

**Paramètre dynamique** : la variable Python `seuil_stock = 20` détermine le filtre.

**Règles de catégorisation** (identiques aux TPs 3.4 et 3.5) :

| Catégorie       | Condition de prix               |
|------------------|----------------------------------|
| Premium          | `prix > 100`                     |
| Standard         | `prix > 30` et `prix <= 100`     |
| Entrée de gamme  | `prix <= 30`                     |

**Rapport attendu en sortie** :

```
==================================================================
        Produits à réapprovisionner (stock <= 20)
==================================================================
Tapis      | Entrée de gamme  |   9.90 € | stock= 0 | CA =    0.00 €
Écran      | Premium          | 199.00 € | stock= 3 | CA =  597.00 €
Clavier    | Standard         |  79.00 € | stock= 8 | CA =  632.00 €
Casque     | Standard         |  49.90 € | stock=12 | CA =  598.80 €
==================================================================
Total : 4 produits concernés
```

Tri par **stock croissant** (du plus urgent au moins urgent).

---

**Étapes guidées** (10 minutes)

1. **Préparer les imports et la variable de filtre** :
   ```python
   import sqlite3
   seuil_stock = 20
   ```

2. **Ouvrir la connexion vers `catalogue.db`** et activer `conn.row_factory = sqlite3.Row` pour pouvoir accéder aux colonnes par nom dans la suite.

3. **Créer le curseur** avec `conn.cursor()`.

4. **Écrire la requête SQL** dans un triple-quoted string. Elle doit :
   - Sélectionner `nom`, `prix`, `stock`
   - Calculer `categorie` en SQL avec un `CASE WHEN ... THEN ... ELSE ... END AS categorie`
   - Calculer `ca_potentiel` avec `prix * stock AS ca_potentiel`
   - Filtrer avec `WHERE stock <= ?` (paramètre `?`, **jamais de f-string**)
   - Trier avec `ORDER BY stock ASC`

5. **Exécuter la requête** avec `cur.execute(sql, (seuil_stock,))` — n'oublie pas la **virgule traînante** dans le tuple à un élément.

6. **Récupérer le résultat** avec `cur.fetchall()` et le stocker dans une variable (ne pas l'appeler 2 fois).

7. **Afficher l'en-tête** :
   - une barre de `=` (utiliser `"=" * 66`)
   - le titre centré : `f"{titre:^66}"` où titre est construit avec `f"Produits à réapprovisionner (stock <= {seuil_stock})"`
   - une autre barre de `=`

8. **Boucler sur les résultats** et afficher chaque ligne en accédant aux colonnes **par nom** (`ligne['nom']`, `ligne['categorie']`, etc.) — c'est le bénéfice de `row_factory`.

9. **Afficher la ligne de fermeture** (une barre de `=`) **puis le total** au format `Total : N produits concernés` (utiliser `len(resultats)` pour N).

10. **Fermer la connexion** avec `conn.close()`.

**Indices pour les specifiers de largeur**

| Colonne     | Format            |
|--------------|-------------------|
| Nom          | `{:<10}`           |
| Catégorie    | `{:<16}`           |
| Prix         | `{:>6.2f}`         |
| Stock        | `{:>2}`            |
| CA potentiel | `{:>7.2f}`         |

**Patterns d'aide-mémoire**

- Triple-quotes `"""..."""` pour la requête SQL multi-ligne (cf. 3.6.3)
- Paramètre `?` + tuple `(seuil_stock,)` (cf. 3.6.6)
- `conn.row_factory = sqlite3.Row` **avant** le curseur (cf. 3.6.7)
- `AS categorie` et `AS ca_potentiel` pour pouvoir écrire `ligne['categorie']` (cf. 3.6.7)


````python
import sqlite3

# Étape 1 — Paramètre de filtre
seuil_stock = 20

# Étape 2 — Ouverture + row_factory (à activer AVANT le curseur)
conn = sqlite3.connect("catalogue.db")
conn.row_factory = sqlite3.Row

# Étape 3 — Création du curseur
cur = conn.cursor()

# Étape 4-5 — Requête SQL + exécution avec paramètre ?
cur.execute("""
    SELECT
        nom,
        prix,
        stock,
        CASE
            WHEN prix > 100 THEN 'Premium'
            WHEN prix > 30  THEN 'Standard'
            ELSE                 'Entrée de gamme'
        END           AS categorie,
        prix * stock  AS ca_potentiel
    FROM produits
    WHERE stock <= ?
    ORDER BY stock ASC
""", (seuil_stock,))

# Étape 6 — Récupération du résultat dans une variable (pas réutiliser cur.fetchall() !)
resultats = cur.fetchall()

# Étape 7 — Affichage de l'en-tête
LARGEUR = 66
titre = f"Produits à réapprovisionner (stock <= {seuil_stock})"

print("=" * LARGEUR)
print(f"{titre:^{LARGEUR}}")
print("=" * LARGEUR)

# Étape 8 — Boucle d'affichage (accès aux colonnes par NOM grâce au row_factory)
for ligne in resultats:
    print(
        f"{ligne['nom']:<10} | "
        f"{ligne['categorie']:<16} | "
        f"{ligne['prix']:>6.2f} € | "
        f"stock={ligne['stock']:>2} | "
        f"CA = {ligne['ca_potentiel']:>7.2f} €"
    )

# Étape 9 — Ligne de fermeture + total
print("=" * LARGEUR)
print(f"Total : {len(resultats)} produits concernés")

# Étape 10 — Fermeture
conn.close()
````

### 3.7.1 Pourquoi pandas — le grand révélateur

#### Trois raisons concrètes

**1. pandas est LE pivot du data analyste Python**

C'est la bibliothèque la plus utilisée du métier — toutes années confondues, tous secteurs confondus. La maîtriser ouvre directement la porte à **tout l'écosystème data** :

- **numpy** pour le calcul scientifique (pandas est construit dessus)

**2. Un DataFrame = un tableau Excel piloté par Python**

C'est l'analogie qui parle immédiatement. Lignes, colonnes nommées, sélection, filtrage, agrégation, formules dérivées — **tout ce que vous faites dans Excel**, mais :

- **scriptable** (un script reproductible à la place de macros bricolées)
- **performant** sur des **millions de lignes** (Excel cale à 1 million)
- **connecté** à toutes les sources (BDD, API, CSV, JSON, Parquet…)
- **traçable** (chaque transformation est du code lisible et versionnable)

**3. pandas ne remplace pas SQL — il l'orchestre**

C'est important : pandas **n'enterre pas** ce que vous venez de voir au chapitre 3.6. Au contraire, le motif reste identique :

```
[SQL fait l'agrégat lourd]  →  pandas reçoit le résultat  →  pandas continue le travail
```

La ligne magique `pd.read_sql_query("SELECT ...", conn)` est la **passerelle directe** entre 3.6 et 3.7. Vous écrivez votre requête SQL comme avant, et pandas vous rend un **DataFrame** directement utilisable.

#### Ce que vous allez gagner concrètement

| Charger une table SQL               | ~5 lignes (`connect`, `cursor`, `execute`, `fetchall`) | **1 ligne** (`pd.read_sql_query(...)`)    |
| Filtrer par condition               | requête SQL + boucle d'affichage                       | **1 expression** (`df[df["prix"] > 30]`)  |
| Agréger par catégorie               | `CASE WHEN` + `GROUP BY` + boucle Python              | **2 lignes** (`df.groupby(...).agg(...)`) |
| Ajouter une colonne calculée        | boucle Python ou `UPDATE` SQL                          | **1 ligne** (`df["ca"] = df["prix"] * df["stock"]`) |

La démonstration concrète arrivera en 3.7.6 : on refera **exactement** le climax de la 3.6.5 (`CASE WHEN + GROUP BY`) en 2 lignes pandas.

### 3.7.2 Les 2 structures clés : `Series` et `DataFrame`

pandas est construit autour de **deux structures fondamentales** :

| Structure       | Dimension | Description                                | Analogie immédiate              |
|-----------------|-----------|--------------------------------------------|----------------------------------|
| **`DataFrame`** | 2D        | tableau à lignes + colonnes nommées        | feuille Excel, table SQL         |
| **`Series`**    | 1D        | une suite de valeurs avec un index         | une colonne Excel, une colonne SQL |

**Règle à mémoriser** : chaque **colonne** d'un DataFrame **EST** une Series. Un DataFrame est donc un dictionnaire ordonné de Series qui partagent le même index (les numéros de ligne).

#### Création d'un DataFrame

La méthode la plus directe : on part d'une **liste de dictionnaires** (la structure vue au chapitre 3.3) et on la passe à `pd.DataFrame(...)` :

```python
import pandas as pd

produits = [
    {"id": 1, "nom": "Casque",  "prix": 49.90, "stock": 12},
    {"id": 2, "nom": "Souris",  "prix": 19.90, "stock": 45},
    # ...
]

df = pd.DataFrame(produits)
```

Le DataFrame s'affiche dans Jupyter sous forme de **tableau HTML formaté** — bien plus lisible qu'un `print()` de liste de dicts brute.

#### ⚠️ Piège n°1 du chapitre : `df["nom"]` vs `df[["nom"]]`

- **`df["nom"]`** (un seul crochet) → renvoie une **`Series`** (1 colonne, format 1D)
- **`df[["nom"]]`** (deux crochets) → renvoie un **`DataFrame`** à une seule colonne (format 2D)

Les deux contiennent les mêmes valeurs, mais ce ne sont **pas les mêmes objets**. La distinction comptera dès qu'on enchaîne des opérations (méthodes différentes, affichage différent, comportement différent).


````python
import pandas as pd

# Création d'un DataFrame depuis une liste de dicts (rappel chapitre 3.3)
produits = [
    {"id": 1, "nom": "Casque",  "prix": 49.90,  "stock": 12},
    {"id": 2, "nom": "Souris",  "prix": 19.90,  "stock": 45},
    {"id": 3, "nom": "Clavier", "prix": 79.00,  "stock":  8},
    {"id": 4, "nom": "Écran",   "prix": 199.00, "stock":  3},
    {"id": 5, "nom": "Webcam",  "prix": 34.90,  "stock": 27},
    {"id": 6, "nom": "Tapis",   "prix":  9.90,  "stock":  0},
]

df = pd.DataFrame(produits)

# Affichage du DataFrame entier
print("--- Le DataFrame complet ---")
print(df)
print()

# =====================================================
# Les 4 réflexes pour inspecter un DataFrame inconnu
# =====================================================
print(f"Shape (lignes, colonnes) : {df.shape}")
print(f"Colonnes                 : {list(df.columns)}")
print(f"Types des colonnes       :")
print(df.dtypes)
print()

print("--- Aperçu : les 3 premières lignes (df.head(3)) ---")
print(df.head(3))
print()

# =====================================================
# PIÈGE N°1 : df["nom"] (Series) vs df[["nom"]] (DataFrame)
# =====================================================
print("--- df['nom'] (UN crochet) → Series ---")
print(df["nom"])
print(f"Type : {type(df['nom']).__name__}")
print()

print("--- df[['nom']] (DEUX crochets) → DataFrame ---")
print(df[["nom"]])
print(f"Type : {type(df[['nom']]).__name__}")
````

### 3.7.3 Charger des données : `read_sql_query`, `read_csv`

pandas peut lire ses données depuis presque n'importe quelle source : bases SQL, fichiers CSV, Excel, JSON, Parquet, API web… avec une méthode dédiée à chaque format.

#### `pd.read_sql_query` — la passerelle directe avec le chapitre 3.6

C'est la ligne qui annule presque tout le code écrit dans le chapitre 3.6.

Au chapitre 3.6, charger une table SQL prenait :

```python
import sqlite3
conn = sqlite3.connect("catalogue.db")
conn.row_factory = sqlite3.Row
cur = conn.cursor()
cur.execute("SELECT * FROM produits")
lignes = cur.fetchall()
# puis boucle d'affichage, conversion en dict, etc.
```

Avec pandas, ça devient :

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("catalogue.db")
df = pd.read_sql_query("SELECT * FROM produits", conn)
conn.close()
```

→ `df` est directement un **DataFrame** prêt à l'emploi : noms de colonnes inférés, types détectés, index automatique. Le motif `cursor → execute → fetchall → boucle` est entièrement absorbé par `read_sql_query`.

⚠️ **À noter** : on continue à passer une **connexion sqlite3** — pandas l'utilise pour exécuter la requête. La connexion à la BDD reste votre responsabilité (ouverture, fermeture). pandas n'orchestre que la lecture du résultat.

#### Autres méthodes de chargement à connaître

| Méthode                  | Source                                              |
|---------------------------|------------------------------------------------------|
| `pd.read_csv(...)`        | fichier CSV (le plus courant en data)               |
| `pd.read_excel(...)`      | fichier Excel `.xlsx`                               |
| `pd.read_json(...)`       | fichier JSON ou réponse d'API web                   |
| `pd.read_parquet(...)`    | fichier Parquet (format colonne haute performance)  |
| `pd.read_sql_query(...)`  | requête SQL sur une connexion BDD                   |

Toutes ont la **même philosophie** : *une ligne pour charger, un DataFrame en sortie*. On découvrira `read_csv` concrètement dans le mini-projet 3.8.


````python
import sqlite3
import pandas as pd

# Connexion à la base SQLite créée au chapitre 3.6.3
conn = sqlite3.connect("catalogue.db")

# UNE seule ligne pour charger toute la table dans un DataFrame
df = pd.read_sql_query("SELECT * FROM produits", conn)

# Affichage du résultat
print("--- DataFrame chargé depuis SQLite ---")
print(df)
print()
print(f"Shape : {df.shape}")
print(f"Type  : {type(df).__name__}")

# =====================================================
# Tu peux passer N'IMPORTE QUELLE requête SQL
# (pas seulement un SELECT *)
# Exemple : filtrage + tri faits côté SQL, pandas reçoit le résultat final
# =====================================================
df_filtre = pd.read_sql_query("""
    SELECT nom, prix, stock
    FROM produits
    WHERE stock > 0
    ORDER BY prix DESC
""", conn)

print()
print("--- Avec filtrage et tri SQL inclus dans la requête ---")
print(df_filtre)

# Fermeture à la fin, une seule fois, après toutes les requêtes
conn.close()
````

### 3.7.4 Sélectionner colonnes et lignes — l'équivalent du `SELECT`

Une fois le DataFrame en main, on veut souvent **n'en garder qu'une partie** : certaines colonnes, certaines lignes, ou un croisement précis. C'est l'équivalent direct du `SELECT col1, col2 FROM ... WHERE ... LIMIT ...` en SQL.

#### Sélectionner des colonnes

| Objectif                                    | Syntaxe pandas         | Équivalent SQL                    |
|---------------------------------------------|------------------------|------------------------------------|
| Une seule colonne (résultat = Series)       | `df["nom"]`            | `SELECT nom FROM produits`         |
| Plusieurs colonnes (résultat = DataFrame)   | `df[["nom", "prix"]]`  | `SELECT nom, prix FROM produits`   |

⚠️ **Rappel du piège n°1 (vu en 3.7.2)** : `df["col"]` (un crochet) = Series, `df[["col"]]` (deux crochets) = DataFrame. Pour récupérer **plusieurs colonnes**, on passe **une liste de noms** entre crochets supplémentaires.

#### Sélectionner des lignes : `.iloc` vs `.loc`

C'est LA distinction à comprendre une fois pour toutes.

| Méthode     | Type d'accès                   | Analogie SQL                                       |
|-------------|--------------------------------|----------------------------------------------------|
| **`.iloc`** | par **position** (entier 0-based) | `LIMIT 1 OFFSET N` — accès positionnel             |
| **`.loc`**  | par **label** (valeur d'index)    | `WHERE id = 3` — accès par identifiant logique     |

**Tant qu'on n'a pas redéfini l'index**, les positions et les labels coïncident (0, 1, 2, 3…). Mais dès qu'on appelle `df.set_index("id")`, `.loc` cherchera dans la colonne `id`, et `.iloc` continuera à compter à partir de 0. C'est là que la différence devient critique.

#### Syntaxes courantes

```python
df.iloc[0]           # 1ère ligne (Series)
df.iloc[0:3]         # 3 premières lignes (DataFrame) — fin exclue, cf 3.3.2
df.iloc[-1]          # dernière ligne
df.iloc[0, 2]        # une cellule : ligne 0, colonne 2

df.loc[2]            # ligne dont l'index vaut 2 (Series)
df.loc[2, "nom"]     # une cellule : ligne d'index 2, colonne "nom"
df.loc[[1, 3, 5]]    # plusieurs lignes par leurs labels (DataFrame)
```


````python
import sqlite3
import pandas as pd

# On recharge le DataFrame depuis SQLite (cellule autonome, rejouable)
conn = sqlite3.connect("catalogue.db")
df = pd.read_sql_query("SELECT * FROM produits", conn)
conn.close()

# =====================================================
# 1) Sélection de COLONNES
# =====================================================
print("--- 1 colonne (Series) : df['nom'] ---")
print(df["nom"])
print()

print("--- 2 colonnes (DataFrame) : df[['nom', 'prix']] ---")
print(df[["nom", "prix"]])
print()

# =====================================================
# 2) Sélection de LIGNES par POSITION avec .iloc
# =====================================================
print("--- Première ligne : df.iloc[0] ---")
print(df.iloc[0])
print()

print("--- 3 premières lignes : df.iloc[0:3] ---")
print(df.iloc[0:3])
print()

print("--- Dernière ligne : df.iloc[-1] ---")
print(df.iloc[-1])
print()

# =====================================================
# 3) Sélection de LIGNES par LABEL avec .loc
# (avec l'index par défaut 0, 1, 2..., .loc et .iloc donnent le même résultat)
# =====================================================
print("--- Ligne d'index 2 : df.loc[2] ---")
print(df.loc[2])
print()

print("--- 3 lignes par leurs labels : df.loc[[1, 3, 5]] ---")
print(df.loc[[1, 3, 5]])
print()

# =====================================================
# 4) Accès à UNE CELLULE précise (croisement ligne × colonne)
# =====================================================
print(f"Prix en ligne 0 (par position) : {df.iloc[0, 2]}")     # ligne 0, colonne 2 (prix)
print(f"Prix en ligne 0 (par label)    : {df.loc[0, 'prix']}")  # ligne d'index 0, colonne 'prix'
print()

# =====================================================
# 5) Démonstration de la VRAIE différence .iloc vs .loc
# Après set_index('id'), l'index n'est plus 0, 1, 2... mais 1, 2, 3... (les vrais IDs)
# =====================================================
df_indexe = df.set_index("id")
print("--- Après set_index('id'), l'index est désormais la colonne id ---")
print(df_indexe)
print()

print("--- df_indexe.iloc[0] : la 1ère ligne (Casque, position 0) ---")
print(df_indexe.iloc[0])
print()

print("--- df_indexe.loc[4] : la ligne dont l'id vaut 4 (Écran) ---")
print(df_indexe.loc[4])
````

### 3.7.5 Filtrer avec le *boolean indexing* — l'équivalent du `WHERE`

Le **boolean indexing** est la façon pandas de filtrer des lignes selon une condition. C'est l'équivalent direct du `WHERE` SQL — mais la syntaxe peut sembler étrange au premier abord.

#### Le principe en 2 étapes

**Étape 1** : une comparaison sur une colonne (Series) produit une **Series de `True` / `False`**, une valeur par ligne :

```python
df["prix"] > 30
# 0     True    (Casque, 49.90 > 30 → True)
# 1    False    (Souris, 19.90 → False)
# 2     True    (Clavier, 79.00)
# 3     True    (Écran, 199.00)
# 4     True    (Webcam, 34.90)
# 5    False    (Tapis, 9.90)
```

**Étape 2** : on **injecte cette Series de booléens dans `df[...]`**, et pandas garde uniquement les lignes où c'est `True` :

```python
df[df["prix"] > 30]
# Garde Casque, Clavier, Écran, Webcam — supprime Souris et Tapis
```

C'est l'équivalent direct de `SELECT * FROM produits WHERE prix > 30`.

#### Conditions multiples

| Sens   | Opérateur pandas | Équivalent SQL |
|--------|-------------------|-----------------|
| ET     | **`&`**           | `AND`           |
| OU     | **`\|`**          | `OR`            |
| NON    | **`~`**           | `NOT`           |

⚠️ **Piège n°2 du chapitre** : avec plusieurs conditions, **les parenthèses sont obligatoires** autour de chaque sous-condition (à cause de la précédence des opérateurs Python).

```python
# ❌ Ne marche PAS — provoque une erreur de type
df[df["prix"] > 30 & df["stock"] < 10]

# ✅ Marche — chaque condition est entre parenthèses
df[(df["prix"] > 30) & (df["stock"] < 10)]
```

Et **ne pas utiliser** `and`, `or`, `not` (les mots Python) — ça plante. Toujours `&`, `|`, `~`.

#### Filtrer par appartenance à une liste (équivalent `WHERE IN`)

Pour vérifier si une colonne **prend une valeur parmi plusieurs**, on utilise `.isin([...])` — plus lisible qu'un enchaînement de `|` :

```python
df[df["nom"].isin(["Casque", "Écran"])]
# Équivalent SQL : SELECT * FROM produits WHERE nom IN ('Casque', 'Écran')
```


````python
import sqlite3
import pandas as pd

# Rechargement (cellule autonome)
conn = sqlite3.connect("catalogue.db")
df = pd.read_sql_query("SELECT * FROM produits", conn)
conn.close()

# =====================================================
# 1) Le mécanisme : une comparaison produit une Series de True/False
# =====================================================
print("--- df['prix'] > 30 : produit une Series de True/False ---")
print(df["prix"] > 30)
print()

# =====================================================
# 2) On injecte cette Series dans df[...] → filtrage
# =====================================================
print("--- df[df['prix'] > 30] : on garde les lignes où c'est True ---")
print(df[df["prix"] > 30])
print()

# =====================================================
# 3) Conditions multiples — PARENTHÈSES OBLIGATOIRES
# =====================================================
print("--- prix > 30 ET stock > 10 ---")
print(df[(df["prix"] > 30) & (df["stock"] > 10)])
print()

print("--- prix > 100 OU stock == 0 ---")
print(df[(df["prix"] > 100) | (df["stock"] == 0)])
print()

print("--- Négation avec ~ : NON premium (= prix <= 100) ---")
print(df[~(df["prix"] > 100)])
print()

# =====================================================
# 4) Filtrage par appartenance à une liste : .isin()
# =====================================================
print("--- nom IN ('Casque', 'Écran', 'Tapis') ---")
print(df[df["nom"].isin(["Casque", "Écran", "Tapis"])])
print()

# =====================================================
# 5) Combiner filtre + sélection de colonne
# =====================================================
print("--- Noms des produits avec stock > 20 ---")
print(df[df["stock"] > 20]["nom"])
print()

# Variante équivalente avec .loc (plus explicite, plus pythonique)
print("--- Idem avec .loc : df.loc[condition, 'colonne'] ---")
print(df.loc[df["stock"] > 20, "nom"])
````

### 3.7.6 Agréger avec `groupby` — le grand révélateur

`groupby` est l'équivalent direct du `GROUP BY` SQL. C'est aussi **LA fonction la plus puissante** de pandas — celle que vous utiliserez le plus au quotidien.

#### Le principe en 3 étapes (split – apply – combine)

1. **Split** : pandas découpe le DataFrame en sous-groupes selon la valeur d'une colonne
2. **Apply** : pandas applique une fonction d'agrégation à chaque sous-groupe (`sum`, `mean`, `count`…)
3. **Combine** : pandas recolle les résultats en un nouveau DataFrame

```python
df.groupby("categorie")["prix"].sum()
# → SELECT categorie, SUM(prix) FROM produits GROUP BY categorie
```

#### ⚠️ Piège n°3 du chapitre : `groupby` seul ne renvoie rien d'affichable

```python
df.groupby("categorie")
# → <pandas.core.groupby.DataFrameGroupBy object at 0x...>
#   ↑ un objet intermédiaire, PAS un DataFrame
```

`groupby` prépare le découpage mais **n'agrège pas tant qu'on ne lui demande pas une opération**. Il faut **obligatoirement le coupler** avec une fonction d'agrégation :

- `.sum()`, `.mean()`, `.count()`, `.min()`, `.max()` — agrégations classiques
- `.agg(["sum", "mean"])` — plusieurs agrégations d'un coup
- `.agg(nouveau_nom=("col", "fonction"))` — agrégation **nommée** (la forme la plus lisible)

#### Le grand révélateur : refaire le climax du chapitre 3.6 en 2 lignes

Au chapitre 3.6.5, on avait écrit cette requête (≈ 12 lignes de SQL + boucle Python) :

```sql
SELECT
    CASE
        WHEN prix > 100 THEN 'Premium'
        WHEN prix > 30  THEN 'Standard'
        ELSE 'Entrée de gamme'
    END AS categorie,
    COUNT(*)            AS nb_produits,
    SUM(prix * stock)   AS ca_potentiel
FROM produits
GROUP BY categorie
ORDER BY ca_potentiel DESC
```

En pandas, ça tient en **3 instructions courtes** :

1. Créer la colonne `categorie` à partir de la fonction `categoriser()` (réutilisée depuis le chapitre 3.5)
2. Créer la colonne `ca_potentiel` (multiplication directe `prix * stock`)
3. `groupby("categorie").agg(...)` + tri

Démonstration dans la cellule code juste en dessous.


````python
import sqlite3
import pandas as pd

# Rechargement (cellule autonome)
conn = sqlite3.connect("catalogue.db")
df = pd.read_sql_query("SELECT * FROM produits", conn)
conn.close()

# =====================================================
# 1) Agrégations sur une colonne entière (sans groupby)
# =====================================================
print(f"Stock total        : {df['stock'].sum()}")
print(f"Prix moyen         : {df['prix'].mean():.2f} €")
print(f"Prix max           : {df['prix'].max():.2f} €")
print(f"Nombre de produits : {df['prix'].count()}")
print()

# =====================================================
# 2) Toutes les stats en une commande : describe()
# =====================================================
print("--- df.describe() : statistiques de toutes les colonnes numériques ---")
print(df.describe())
print()

# =====================================================
# 3) groupby SEUL ne renvoie rien d'affichable (piège n°3)
# =====================================================
print("--- df.groupby('nom') seul : un objet intermédiaire, pas un résultat ---")
print(df.groupby("nom"))
print()

# =====================================================
# 4) Préparation : on crée la colonne 'categorie' avec la fonction de 3.5
# =====================================================
def categoriser(prix):
    if prix > 100:
        return "Premium"
    elif prix > 30:
        return "Standard"
    else:
        return "Entrée de gamme"

# UNE ligne pour appliquer la fonction à toute la colonne prix
df["categorie"] = df["prix"].apply(categoriser)

# UNE ligne pour calculer le CA potentiel (multiplication colonne à colonne)
df["ca_potentiel"] = df["prix"] * df["stock"]

print("--- DataFrame enrichi avec 'categorie' et 'ca_potentiel' ---")
print(df)
print()

# =====================================================
# 5) LE CLIMAX : groupby + agrégation nommée
# Équivalent du SELECT CASE WHEN + GROUP BY de la 3.6.5, en 1 instruction pandas
# =====================================================
rapport = (
    df.groupby("categorie")
      .agg(
          nb_produits=("nom", "count"),
          ca_potentiel=("ca_potentiel", "sum"),
      )
      .sort_values("ca_potentiel", ascending=False)
)

print("--- Climax : équivalent du SELECT CASE WHEN + GROUP BY de la 3.6.5 ---")
print(rapport)
````

### 3.7.7 Calculer une colonne dérivée — `df["..."] = ...` et `.apply()`

En analyse data, on enrichit constamment les DataFrames avec des **colonnes calculées** : CA = prix × stock, marge = prix − coût, catégorie selon le prix, prix TTC, etc.

pandas offre **2 façons principales** de créer une colonne dérivée — du plus simple au plus flexible.

#### 1. Arithmétique vectorisée (le plus simple et le plus rapide)

Quand le calcul est une **opération arithmétique simple** sur d'autres colonnes, on l'écrit directement :

```python
df["ca_potentiel"] = df["prix"] * df["stock"]    # colonne × colonne
df["prix_ttc"]     = df["prix"] * 1.20           # colonne × constante
df["marge"]        = df["prix"] - df["cout"]     # colonne − colonne
```

pandas applique l'opération **élément par élément** (on parle de **vectorisation**). C'est **très rapide**, même sur des millions de lignes — pas de boucle Python visible. **À privilégier dès que le calcul est exprimable en arithmétique pure.**

#### 2. `.apply()` avec une fonction

Quand la logique est trop complexe pour de l'arithmétique pure (conditions, transformations…), on passe par **une fonction** qu'on applique à toute la colonne.

```python
def categoriser(prix):
    if prix > 100:
        return "Premium"
    elif prix > 30:
        return "Standard"
    else:
        return "Entrée de gamme"

df["categorie"] = df["prix"].apply(categoriser)
```

→ pandas appelle `categoriser(valeur)` pour **chaque valeur** de la colonne `prix`. C'est l'équivalent direct d'une boucle `for` — mais condensé en une expression.

🎯 **Pont avec le chapitre 3.5** : la fonction `categoriser()` est strictement **celle qu'on a écrite au Mini-TP 3.5**. Pas modifiée, pas adaptée à pandas — réutilisée **telle quelle**. C'est la démonstration concrète du bénéfice DRY vu en 3.5.1.

#### Récap : quelle méthode choisir ?

| Type d'opération                                  | Méthode à privilégier                       |
|---------------------------------------------------|----------------------------------------------|
| Arithmétique simple (`+`, `-`, `*`, `/`)          | **Arithmétique directe** (méthode 1)         |
| Logique conditionnelle (`if/elif/else`)           | **`.apply(fonction)`** (méthode 2)           |


````python
import sqlite3
import pandas as pd

# Rechargement (cellule autonome)
conn = sqlite3.connect("catalogue.db")
df = pd.read_sql_query("SELECT * FROM produits", conn)
conn.close()

# =====================================================
# Méthode 1 — Arithmétique directe (la plus simple)
# =====================================================
df["ca_potentiel"] = df["prix"] * df["stock"]    # colonne × colonne
df["prix_ttc"]     = df["prix"] * 1.20           # colonne × constante

print("--- Méthode 1 : arithmétique directe ---")
print(df[["nom", "prix", "stock", "ca_potentiel", "prix_ttc"]])

print()

# =====================================================
# Méthode 2 — .apply() avec une fonction nommée (rappel 3.5)
# =====================================================
def categoriser(prix):
    if prix > 100:
        return "Premium"
    elif prix > 30:
        return "Standard"
    else:
        return "Entrée de gamme"

df["categorie"] = df["prix"].apply(categoriser)

print("--- Méthode 2 : .apply() avec fonction nommée ---")
print(df[["nom", "prix", "categorie"]])
print()

# =====================================================
# Bonus : créer une colonne booléenne (très utilisé en data)
# Une comparaison produit une Series de True/False → directement assignable
# =====================================================
df["en_rupture"] = df["stock"] == 0

print("--- Colonne booléenne 'en_rupture' ---")
print(df[["nom", "stock", "en_rupture"]])
````

### 📋 3.7.8 Récap + grande table de correspondance SQL ↔ pandas

Vous venez d'apprendre les bases de pandas. Vous savez maintenant faire **tout ce que vous faisiez en SQL** — directement dans Python.

#### Les structures à retenir

- **`DataFrame`** = tableau 2D (lignes × colonnes nommées) — équivalent table SQL ou feuille Excel
- **`Series`** = colonne 1D — équivalent colonne SQL
- Chaque colonne d'un DataFrame **EST** une Series

---

#### Grande table de correspondance SQL ↔ pandas

| Opération SQL                                              | Équivalent pandas                                                  |
|------------------------------------------------------------|---------------------------------------------------------------------|
| `SELECT * FROM table`                                      | `df`                                                                |
| `SELECT col FROM table`                                    | `df["col"]` *(renvoie une Series)*                                  |
| `SELECT col1, col2 FROM table`                             | `df[["col1", "col2"]]`                                              |
| `SELECT * FROM table LIMIT 5`                              | `df.head(5)` ou `df.iloc[:5]`                                       |
| `SELECT * FROM table WHERE col > 30`                       | `df[df["col"] > 30]`                                                |
| `SELECT * FROM table WHERE col1 > 30 AND col2 < 10`        | `df[(df["col1"] > 30) & (df["col2"] < 10)]`                         |
| `SELECT * FROM table WHERE col IN ('A', 'B')`              | `df[df["col"].isin(["A", "B"])]`                                    |
| `SELECT * FROM table ORDER BY col DESC`                    | `df.sort_values("col", ascending=False)`                            |
| `SELECT COUNT(*) FROM table`                               | `len(df)` ou `df.shape[0]`                                          |
| `SELECT AVG(col) FROM table`                               | `df["col"].mean()`                                                  |
| `SELECT SUM(col) FROM table`                               | `df["col"].sum()`                                                   |
| `SELECT MIN(col), MAX(col) FROM table`                     | `df["col"].min()`, `df["col"].max()`                                |
| `SELECT col, COUNT(*) FROM table GROUP BY col`             | `df.groupby("col").size()`                                          |
| `SELECT col, SUM(autre) FROM table GROUP BY col`           | `df.groupby("col")["autre"].sum()`                                  |
| `SELECT DISTINCT col FROM table`                           | `df["col"].unique()` ou `df.drop_duplicates(subset="col")`          |
| `ALTER TABLE + UPDATE` (colonne calculée arithmétique)     | `df["nouvelle"] = df["x"] * df["y"]`                                |
| `CASE WHEN ... END` (catégorisation conditionnelle)        | `df["cat"] = df["x"].apply(fonction_categoriser)`                   |
| `SELECT * FROM t1 JOIN t2 ON t1.id = t2.id`                | `pd.merge(t1, t2, on="id")`                                         |

---

#### Les 3 pièges du chapitre

1. **`df["col"]` (Series) vs `df[["col"]]` (DataFrame)** — un crochet vs deux crochets. Pour plusieurs colonnes, **toujours** une **liste Python** à l'intérieur des crochets pandas.

2. **Conditions multiples : parenthèses obligatoires** — `df[(df["a"] > 1) & (df["b"] < 2)]`. Sans les parenthèses autour de chaque sous-condition, ça plante. Et utiliser `&` `|` `~`, **pas** `and` `or` `not`.

3. **`df.groupby("col")` seul ne renvoie rien d'affichable** — il faut **toujours** le coupler avec une fonction d'agrégation : `.sum()`, `.mean()`, `.count()`, `.agg(...)`.

---

#### La règle de partage SQL ↔ pandas

> **SQL pour le pré-traitement lourd** : filtrer, joindre, agréger côté base **avant** de tirer les données
>
> **pandas pour l'analyse interactive** : transformations en chaîne, visualisations, exports, combinaisons de sources

Vous écrirez très souvent ce pattern :

```python
df = pd.read_sql_query("""
    SELECT ... [requête métier filtrée et agrégée]
""", conn)

# Puis enchaînement pandas pour les transformations détaillées
df["nouvelle_col"] = df["x"] * df["y"]
df.groupby("cat").agg(...)
# etc.
```

### 3.8.2 Q1 — Quel CA total a généré le trimestre ?

C'est la question la plus simple, et celle qu'on traite **pas à pas** ensemble pour ancrer le pattern de base que vous réutiliserez pour toutes les questions suivantes.

#### L'énoncé

> *"Quel est le chiffre d'affaires total généré par les ventes du trimestre ?"*

**Attention au piège métier** : on parle bien des **ventes**, pas des retours. Un retour annule une vente — il faut donc **exclure les lignes** où `statut == "Retour"`.

#### La démarche en 3 étapes

| Étape | Opération                                       | Outil pandas (rappel)                      |
|-------|--------------------------------------------------|---------------------------------------------|
| 1     | **Filtrer** : ne garder que les ventes           | Boolean indexing (3.7.5)                    |
| 2     | **Calculer le CA** de chaque ligne (`prix × quantité`) | Colonne dérivée par arithmétique (3.7.7) |
| 3     | **Sommer** tous les CA                           | `.sum()` (3.7.6)                            |

C'est le **pattern fondamental** : filtre → colonne calculée → agrégation. On le retrouve absolument partout en analyse data.

La cellule de code suivante implémente ces 3 étapes. Suivez les commentaires pour faire le lien.


````python
import pandas as pd

# Rechargement du CSV (cellule autonome)
df = pd.read_csv("ventes_retours.csv")

# =====================================================
# Étape 1 — Filtrer : ne garder que les lignes "Vente"
# =====================================================
df_ventes = df[df["statut"] == "Vente"].copy()
# Le .copy() évite un avertissement pandas quand on ajoute une colonne juste après

# =====================================================
# Étape 2 — Créer la colonne CA = prix unitaire × quantité
# =====================================================
df_ventes["ca"] = df_ventes["prix_unitaire"] * df_ventes["quantite"]

# =====================================================
# Étape 3 — Sommer la colonne CA pour obtenir le total
# =====================================================
ca_total = df_ventes["ca"].sum()

# Affichage du résultat
print(f"CA total du trimestre (ventes uniquement) : {ca_total:.2f} €")
print()
print(f"Nombre de ventes  : {len(df_ventes)}")
print(f"Nombre de retours : {len(df) - len(df_ventes)} (exclus du calcul)")
````

### 3.8.3 Q2 — Quels sont les 5 produits qui rapportent le plus ?

Votre responsable veut connaître les **locomotives du catalogue** : les cinq produits qui génèrent le plus de chiffre d'affaires sur le trimestre.

#### L'énoncé

> *"Donne-moi le top 5 des produits par chiffre d'affaires, ventes uniquement. Et tant qu'à faire, montre aussi le nombre de ventes de chacun — je veux voir si les gros CA sont portés par le volume ou par le prix."*

#### La démarche

On reprend **exactement le pattern de Q1** — filtre → colonne calculée → agrégation — mais cette fois l'agrégation ne donne plus un seul chiffre : on veut **un total par produit**. C'est le moment de sortir le `groupby` (l'équivalent pandas du `GROUP BY` SQL).

| Étape | Opération                                              | Outil pandas                          |
|-------|--------------------------------------------------------|---------------------------------------|
| 1     | **Filtrer** : ventes uniquement                        | Boolean indexing (comme Q1)           |
| 2     | **Créer la colonne `ca`** (`prix × quantité`)          | Colonne dérivée (comme Q1)            |
| 3     | **Grouper par `produit`** : somme du CA + nb de ventes | `groupby` + `.agg()`                   |
| 4     | **Trier** par CA décroissant et garder les 5 premiers  | `.sort_values()` + `.head(5)`         |

> 💡 **Rappel SQL** : `GROUP BY produit` + `SUM(ca)` + `COUNT(*)` + `ORDER BY ca DESC LIMIT 5`. Vous savez déjà raisonner comme ça — il ne reste qu'à le traduire en pandas.

La cellule suivante implémente ces 4 étapes. Suivez les commentaires.


````python
import pandas as pd

# Rechargement du CSV (cellule autonome)
df = pd.read_csv("ventes_retours.csv")

# =====================================================
# Étape 1 — Filtrer : ventes uniquement (comme en Q1)
# =====================================================
df_ventes = df[df["statut"] == "Vente"].copy()

# =====================================================
# Étape 2 — Colonne CA = prix unitaire × quantité (comme en Q1)
# =====================================================
df_ventes["ca"] = df_ventes["prix_unitaire"] * df_ventes["quantite"]

# =====================================================
# Étape 3 — Grouper par produit : somme du CA + nombre de ventes
# .agg() permet de calculer plusieurs mesures d'un coup et de nommer
# les colonnes résultat (ca_total, nb_ventes)
# =====================================================
par_produit = df_ventes.groupby("produit").agg(
    ca_total=("ca", "sum"),
    nb_ventes=("ca", "count"),
)

# =====================================================
# Étape 4 — Trier par CA décroissant et garder le top 5
# =====================================================
top5 = par_produit.sort_values("ca_total", ascending=False).head(5)

print("--- Top 5 produits par chiffre d'affaires (ventes uniquement) ---")
print(top5)
````

### 3.8.4 Q3 — Quel est le taux de retour par catégorie ?

Les retours coûtent cher (logistique inverse, remise en stock, parfois destruction). Votre responsable veut savoir **quelles catégories sont les plus retournées** pour cibler les actions correctives.

#### L'énoncé

> *"Calcule-moi le taux de retour par catégorie de produit. Je veux voir, pour Premium, Standard et Entrée de gamme, le pourcentage de commandes qui finissent en retour."*

#### Un changement important par rapport à Q1 et Q2

Jusqu'ici, on **filtrait sur les ventes** dès la première étape. **Ici, surtout pas !** Un taux de retour, c'est :

taux de retour = nombre de retours ÷ nombre total de commandes


Il nous faut donc **garder toutes les lignes** — ventes *et* retours — car les retours sont au numérateur et le total (ventes + retours) au dénominateur. Si on filtrait les ventes d'entrée de jeu, on perdrait justement ce qu'on veut compter.

#### La démarche

| Étape | Opération                                                  | Outil pandas                          |
|-------|-------------------------------------------------------------|---------------------------------------|
| 1     | **Créer une colonne booléenne** `est_retour` (`True`/`False`) | Comparaison → Series booléenne (3.7.7) |
| 2     | **Grouper par `categorie`** : nombre total + nombre de retours | `groupby` + `.agg()`                   |
| 3     | **Calculer le taux** = retours ÷ total × 100               | Colonne dérivée + `.round(1)`         |
| 4     | **Trier** par taux décroissant                            | `.sort_values()`                      |

> 💡 **Rappel SQL** : c'est l'équivalent de `SUM(CASE WHEN statut='Retour' THEN 1 ELSE 0 END) / COUNT(*)` groupé par catégorie. En pandas, le `CASE WHEN` devient une simple **colonne booléenne**.

La cellule suivante implémente ces 4 étapes.


````python
import pandas as pd

# Rechargement du CSV (cellule autonome)
df = pd.read_csv("ventes_retours.csv")

# =====================================================
# Étape 1 — Colonne booléenne : True si la ligne est un retour
# Une comparaison renvoie une Series de True/False, directement assignable
# =====================================================
df["est_retour"] = df["statut"] == "Retour"

# =====================================================
# Étape 2 — Grouper par catégorie :
#   - nb_total   = nombre de lignes du groupe  -> count
#   - nb_retours = nombre de True              -> sum (True vaut 1, False vaut 0)
# =====================================================
par_cat = df.groupby("categorie").agg(
    nb_total=("est_retour", "count"),
    nb_retours=("est_retour", "sum"),
)

# =====================================================
# Étape 3 — Taux de retour en % (arrondi à 1 décimale)
# =====================================================
par_cat["taux_retour_%"] = (par_cat["nb_retours"] / par_cat["nb_total"] * 100).round(1)

# =====================================================
# Étape 4 — Trier par taux décroissant (la catégorie la plus retournée en haut)
# =====================================================
par_cat = par_cat.sort_values("taux_retour_%", ascending=False)

print("--- Taux de retour par catégorie ---")
print(par_cat)
````

### 3.8.5 Q4 — Quelles régions génèrent le plus de chiffre d'affaires ?

Le service marketing prépare ses campagnes régionales et la logistique veut optimiser ses entrepôts. Tous deux ont besoin de savoir **où se concentre le chiffre d'affaires**.

#### L'énoncé

> *"Classe-moi les régions par chiffre d'affaires, ventes uniquement. Et donne-moi aussi le poids de chaque région en pourcentage du CA total — je veux savoir à quel point c'est concentré."*

#### La démarche

Bonne nouvelle : c'est **le pattern de Q2 presque à l'identique**. On groupe par région au lieu de par produit, et on ajoute une **colonne de pourcentage** pour mesurer le poids relatif de chacune.

| Étape | Opération                                              | Outil pandas                          |
|-------|--------------------------------------------------------|---------------------------------------|
| 1     | **Filtrer** : ventes uniquement                        | Boolean indexing (comme Q1/Q2)        |
| 2     | **Créer la colonne `ca`** (`prix × quantité`)          | Colonne dérivée (comme Q1/Q2)         |
| 3     | **Grouper par `client_region`** : somme du CA          | `groupby` + `.agg()`                   |
| 4     | **Ajouter la part en %** : CA région ÷ CA total × 100  | Colonne dérivée à partir d'un total   |
| 5     | **Trier** par CA décroissant                          | `.sort_values()`                      |

> 💡 **Rappel SQL** : `SELECT client_region, SUM(ca) FROM ... WHERE statut='Vente' GROUP BY client_region ORDER BY SUM(ca) DESC`. La part en % se calcule ensuite en divisant par le total général.

La cellule suivante implémente ces 5 étapes.


````python
import pandas as pd

# Rechargement du CSV (cellule autonome)
df = pd.read_csv("ventes_retours.csv")

# =====================================================
# Étape 1 — Filtrer : ventes uniquement (comme Q1/Q2)
# =====================================================
df_ventes = df[df["statut"] == "Vente"].copy()

# =====================================================
# Étape 2 — Colonne CA = prix unitaire × quantité (comme Q1/Q2)
# =====================================================
df_ventes["ca"] = df_ventes["prix_unitaire"] * df_ventes["quantite"]

# =====================================================
# Étape 3 — Grouper par région : somme du CA
# =====================================================
par_region = df_ventes.groupby("client_region").agg(ca_total=("ca", "sum"))

# =====================================================
# Étape 4 — Part de chaque région dans le CA total (en %)
# On calcule d'abord le CA global, puis chaque part
# =====================================================
ca_global = par_region["ca_total"].sum()
par_region["part_%"] = (par_region["ca_total"] / ca_global * 100).round(1)

# =====================================================
# Étape 5 — Trier par CA décroissant (la région la plus forte en haut)
# =====================================================
par_region = par_region.sort_values("ca_total", ascending=False)

print("--- Chiffre d'affaires par région (ventes uniquement) ---")
print(par_region)
````

### 3.8.6 Q5 — Comment le chiffre d'affaires évolue-t-il mois par mois ?

Dernière question, et sans doute la plus stratégique pour le comité de direction : **la tendance**. Le trimestre est-il en croissance, stable, ou en repli ?

#### L'énoncé

> *"Donne-moi le CA mois par mois sur le trimestre, ventes uniquement. Je veux voir la courbe : est-ce qu'on monte, est-ce qu'on stagne ?"*

#### La nouveauté : travailler avec des dates

Jusqu'ici, on groupait par une colonne texte (produit, catégorie, région). Cette fois, il faut grouper par **mois** — or la colonne `date` contient des jours précis (`2026-02-15`). Deux étapes nouvelles :

1. **Convertir** la colonne `date` en vraie date pandas (`datetime`), pas un simple texte. On le fait dès le chargement avec `parse_dates=["date"]`.
2. **Extraire le mois** de chaque date pour pouvoir grouper dessus.

#### La démarche

| Étape | Opération                                              | Outil pandas                          |
|-------|--------------------------------------------------------|---------------------------------------|
| 1     | **Charger en convertissant la date** en datetime       | `read_csv(..., parse_dates=["date"])` |
| 2     | **Filtrer** : ventes uniquement                        | Boolean indexing (comme avant)        |
| 3     | **Créer la colonne `ca`** (`prix × quantité`)          | Colonne dérivée (comme avant)         |
| 4     | **Extraire le mois** (`2026-01`, `2026-02`, …)         | Accesseur `.dt` sur la colonne date   |
| 5     | **Grouper par mois** : somme du CA                     | `groupby` + `.agg()`                   |

> 💡 **Rappel SQL** : l'équivalent de `strftime('%Y-%m', date)` en SQLite, puis `GROUP BY` dessus. En pandas, l'accesseur `.dt` joue ce rôle d'extraction sur les colonnes de type date.

La cellule suivante implémente ces 5 étapes.


````python
import pandas as pd

# =====================================================
# Étape 1 — Charger en convertissant "date" en vraie date (datetime)
# parse_dates transforme le texte "2026-02-15" en objet date manipulable
# =====================================================
df = pd.read_csv("ventes_retours.csv", parse_dates=["date"])

# =====================================================
# Étape 2 — Filtrer : ventes uniquement (comme avant)
# =====================================================
df_ventes = df[df["statut"] == "Vente"].copy()

# =====================================================
# Étape 3 — Colonne CA = prix unitaire × quantité (comme avant)
# =====================================================
df_ventes["ca"] = df_ventes["prix_unitaire"] * df_ventes["quantite"]

# =====================================================
# Étape 4 — Extraire le mois sous la forme "AAAA-MM"
# .dt donne accès aux composants d'une date ; .strftime formate le texte
# =====================================================
df_ventes["mois"] = df_ventes["date"].dt.strftime("%Y-%m")

# =====================================================
# Étape 5 — Grouper par mois : somme du CA
# Les mois "2026-01" < "2026-02" < "2026-03" se trient naturellement
# =====================================================
par_mois = df_ventes.groupby("mois").agg(ca_total=("ca", "sum"))

print("--- Évolution du CA par mois (ventes uniquement) ---")
print(par_mois)
````

Vous disposez maintenant d'une **boîte à outils complète** pour répondre à la grande majorité des questions d'analyse qu'on vous posera au quotidien. La cellule suivante rassemble les cinq réponses en un seul tableau de bord.




````python
import pandas as pd

# =====================================================
# Cellule autonome : on reconstruit les 5 réponses d'un coup.
# Les variables créées ici (top5_produits, taux_retour, ca_region,
# ca_mois, ca_total) seront RÉUTILISÉES par la cellule d'export Excel (3.8.8).
# =====================================================
df = pd.read_csv("ventes_retours.csv", parse_dates=["date"])

# --- Socle commun : ventes + colonne CA (sert à Q1, Q2, Q4, Q5) ---
ventes = df[df["statut"] == "Vente"].copy()
ventes["ca"] = ventes["prix_unitaire"] * ventes["quantite"]

# --- Q1 : CA total ---
ca_total = ventes["ca"].sum()

# --- Q2 : Top 5 produits ---
top5_produits = (
    ventes.groupby("produit")
    .agg(ca_total=("ca", "sum"), nb_ventes=("ca", "count"))
    .sort_values("ca_total", ascending=False)
    .head(5)
)

# --- Q3 : Taux de retour par catégorie (sur TOUTES les lignes) ---
df["est_retour"] = df["statut"] == "Retour"
taux_retour = df.groupby("categorie").agg(
    nb_total=("est_retour", "count"),
    nb_retours=("est_retour", "sum"),
)
taux_retour["taux_%"] = (taux_retour["nb_retours"] / taux_retour["nb_total"] * 100).round(1)
taux_retour = taux_retour.sort_values("taux_%", ascending=False)

# --- Q4 : CA par région ---
ca_region = (
    ventes.groupby("client_region")
    .agg(ca_total=("ca", "sum"))
    .sort_values("ca_total", ascending=False)
)

# --- Q5 : CA par mois ---
ventes["mois"] = ventes["date"].dt.strftime("%Y-%m")
ca_mois = ventes.groupby("mois").agg(ca_total=("ca", "sum"))

# =====================================================
# Affichage du tableau de bord
# =====================================================
print("=" * 55)
print("  TABLEAU DE BORD — VENTES Q1 2026 (Cdiscount)")
print("=" * 55)
print()
print(f"CA total du trimestre : {ca_total:.2f} €  ({len(ventes)} ventes)")
print()
print("--- Top 5 produits ---")
print(top5_produits)
print()
print("--- Taux de retour par catégorie ---")
print(taux_retour)
print()
print("--- Top 3 régions ---")
print(ca_region.head(3))
print()
print("--- Évolution mensuelle ---")
print(ca_mois)
````

````python
import pandas as pd
from openpyxl.styles import Font, PatternFill, Alignment

# =====================================================
# /!\ DÉPENDANCE : réutilise les variables de la synthèse (3.8.7) :
#   ca_total, ventes, df, top5_produits, taux_retour, ca_region, ca_mois
# =====================================================

# --- Styles communs à toutes les feuilles (définis une seule fois) ---
EN_TETE_FONT = Font(bold=True, color="FFFFFF", size=12)   # texte blanc, gras
EN_TETE_FILL = PatternFill("solid", fgColor="1F4E78")     # fond bleu foncé
CENTRE       = Alignment(horizontal="center", vertical="center")

def formater_feuille(ws, largeurs, colonnes_euro=()):
    """Applique une mise en forme standard à une feuille Excel.
    - ws            : la feuille (objet openpyxl)
    - largeurs      : dict {lettre_colonne: largeur}, ex {"A": 18, "B": 14}
    - colonnes_euro : lettres des colonnes à afficher en format monétaire €
    """
    # En-tête (ligne 1) : gras blanc sur fond bleu, centré
    for cell in ws[1]:
        cell.font = EN_TETE_FONT
        cell.fill = EN_TETE_FILL
        cell.alignment = CENTRE

    # Largeurs de colonnes
    for lettre, largeur in largeurs.items():
        ws.column_dimensions[lettre].width = largeur

    # Format monétaire (€) sur les colonnes demandées (on saute la ligne 1 d'en-tête)
    for lettre in colonnes_euro:
        for cell in ws[lettre][1:]:
            cell.number_format = "#,##0.00 €"

    # Figer la ligne d'en-tête pour qu'elle reste visible au défilement
    ws.freeze_panes = "A2"


# --- Feuille de synthèse ---
synthese = pd.DataFrame({
    "indicateur": ["CA total (€)", "Nombre de ventes", "Nombre de retours"],
    "valeur":     [round(ca_total, 2), len(ventes), len(df) - len(ventes)],
})

fichier_sortie = "rapport_ventes_Q1_2026_v2.xlsx"

with pd.ExcelWriter(fichier_sortie, engine="openpyxl") as writer:
    # 1. Écriture des données (comme avant)
    synthese.to_excel(writer,      sheet_name="Synthèse",       index=False)
    top5_produits.to_excel(writer, sheet_name="Top produits")
    taux_retour.to_excel(writer,   sheet_name="Taux de retour")
    ca_region.to_excel(writer,     sheet_name="CA par région")
    ca_mois.to_excel(writer,       sheet_name="CA par mois")

    # 2. Mise en forme de chaque feuille via la fonction
    formater_feuille(writer.sheets["Synthèse"],       {"A": 20, "B": 16})
    formater_feuille(writer.sheets["Top produits"],   {"A": 18, "B": 14, "C": 12}, colonnes_euro=["B"])
    formater_feuille(writer.sheets["Taux de retour"], {"A": 18, "B": 12, "C": 12, "D": 10})
    formater_feuille(writer.sheets["CA par région"],  {"A": 28, "B": 14},          colonnes_euro=["B"])
    formater_feuille(writer.sheets["CA par mois"],    {"A": 12, "B": 14},          colonnes_euro=["B"])

print(f"Rapport Excel formaté créé : {fichier_sortie}")
print("5 feuilles stylées : en-têtes en gras, largeurs ajustées, montants en €")
````

### 3.8.9 Mini-défi — lire un export Excel français

Voici **le problème n°1** des data analysts qui passent d'Excel à Python. Vous le rencontrerez dès votre première semaine en poste, garanti.

#### Le scénario

Un collègue vous envoie un fichier exporté depuis **Excel sur un poste Windows français** : `ventes_retours_export_excel.csv`. Ce sont **exactement les mêmes données** que `ventes_retours.csv` — mêmes 150 commandes, mêmes colonnes. Mais le format, lui, est « à la française ».

Vous tentez le réflexe habituel :

```python
df = pd.read_csv("ventes_retours_export_excel.csv")
…et ça plante (ou pire : ça « marche » mais tout est faux). Bienvenue dans le monde réel.

Les 5 pièges du CSV Excel français — vos 5 réflexes à acquérir
Ouvrez le fichier dans un éditeur de texte (Bloc-notes) et observez. Vous allez repérer cinq différences par rapport à un CSV « propre ». Chacune correspond à un paramètre de read_csv à corriger :

#	Le piège	Ce qu'Excel FR fait	Le réflexe pandas
1	Encodage	cp1252 (latin-1), pas UTF-8 → les accents plantent	encoding="cp1252"
2	Séparateur	; au lieu de , → tout dans une seule colonne	sep=";"
3	Décimale	virgule 19,90 au lieu de point 19.90 → prix lus comme texte	decimal=","
4	Dates en texte	02/01/2026 reste une chaîne de caractères	parse_dates=["date"]
5	Ordre jour/mois	02/01/2026 = 2 janvier, PAS 1er février	dayfirst=True
💡 Pourquoi ; comme séparateur ? Justement parce que la virgule sert déjà de séparateur décimal en français. Excel FR ne peut pas utiliser la virgule pour deux choses à la fois, alors il bascule sur le point-virgule. Tout se tient.

un read_csv qui charge ce fichier correctement, de sorte que le DataFrame obtenu soit identique à celui du CSV propre : accents corrects, prix_unitaire numérique, date en vraie date. La solution tient en un seul appel bien paramétré.





````python

import pandas as pd

# =====================================================
# La solution : un seul read_csv, mais 5 paramètres pour
# désamorcer les 5 pièges du format Excel français
# =====================================================
df = pd.read_csv(
    "ventes_retours_export_excel.csv",
    encoding="cp1252",        # piège 1 — accents (latin-1, pas UTF-8)
    sep=";",                  # piège 2 — séparateur point-virgule
    decimal=",",              # piège 3 — décimale en virgule
    parse_dates=["date"],     # piège 4 — convertir la colonne date
    dayfirst=True,            # piège 5 — jour/mois inversés (02/01 = 2 janvier)
)

# --- Vérifications : le DataFrame est-il propre ? ---
print("Shape :", df.shape)
print()
print("Types des colonnes :")
print(df.dtypes)
print()
print("--- 3 premières lignes ---")
print(df.head(3))
````
